## 1. Condicionales `if`, `elif`, `else`

```python
edad = 17

if edad >= 18:
    print("Eres mayor de edad.")
elif edad >= 16:
    print("Puedes votar con autorización.")
else:
    print("Eres menor de edad.")
```

## 2. Bucle `for`

```python
# Imprimir los números del 1 al 5
for i in range(1, 6):
    print(i,end=' ')

cadena = "Hola, mundo!"
for char in cadena:
    print(char,end='')
```

## 3. Bucle `while`

```python
contador = 5

while contador < 5:
    print("Contador:", contador)
    contador += 1

# simular un do-while
contador = 5
while True:
    print("Este bloque se ejecuta al menos una vez.")
    if contador >= 5:
        break
    print("Contador:", contador)
    contador += 1
```

### 5.2. Formas de recibir parámetros en funciones

En Python, las funciones pueden recibir parámetros de varias maneras. A continuación, se explican todas las formas principales:

#### 1. **Parámetros posicionales**
Los parámetros se pasan en el orden en que se definen en la función.

```python
def sumar(a, b):
    return a + b

resultado = sumar(3, 5)  # a=3, b=5
```

#### 2. **Parámetros con valores por defecto**
Puedes asignar valores predeterminados a los parámetros. Si no se pasan, se usan los valores por defecto.




```python
def saludar(nombre="Invitado", saludo="Hola"):
    print(f"{saludo}, {nombre}!")

saludar()  # Usa valores por defecto
saludar("Ana")  # nombre="Ana", saludo por defecto
saludar("Luis", "Buenos días")  # Ambos especificados


```

#### 3. **Parámetros con nombre (keyword arguments)**
Puedes pasar argumentos especificando el nombre del parámetro, lo que permite cambiar el orden.



```python
def dividir(dividendo, divisor):
    return dividendo / divisor

resultado = dividir(divisor=2, dividendo=10)  # Orden cambiado

```

#### 4. **Argumentos variables posicionales (*args)**
Usa `*args` para aceptar un número variable de argumentos posicionales, que se almacenan en una tupla.



```python
def sumar_todos(*numeros):
    return sum(numeros)

total = sumar_todos(1, 2, 3, 4)  # numeros = (1, 2, 3, 4)
```

#### 5. **Argumentos variables con nombre (**kwargs)**
Usa `**kwargs` para aceptar un número variable de argumentos con nombre, que se almacenan en un diccionario.



```python
def imprimir_info(**info):
    for clave, valor in info.items():
        print(f"{clave}: {valor}")

imprimir_info(nombre="Ana", edad=25, ciudad="Madrid")

```

#### 6. **Parámetros solo posicionales (Python 3.8+)**
Usa `/` para indicar que ciertos parámetros solo pueden pasarse posicionalmente.

```python
def func(a, b, /, c, d):
    return a + b + c + d

resultado = func(1, 2, c=3, d=4)  # a y b solo posicionales
```

#### 7. **Parámetros solo con nombre**
Usa `*` para indicar que los parámetros siguientes solo pueden pasarse con nombre.



```python
def func(a, *, b, c):
    return a + b + c

resultado = func(1, b=2, c=3)  # b y c solo con nombre
```

#### 8. **Combinación de todas las formas**
Puedes combinar todas estas formas en una sola función.



```python
def funcion_completa(a, b=10, *args, c, d=20, **kwargs):
    print(f"a={a}, b={b}, args={args}, c={c}, d={d}, kwargs={kwargs}")

funcion_completa(1, 2, 3, 4, c=5, e=6, f=7)
```

Estas formas permiten flexibilidad al definir y llamar funciones en Python.

#### 9. Match-case
La estructura match-case fue introducida en Python 3.10 como una alternativa al clásico switch-case de otros lenguajes de programación. Esta herramienta permite realizar concordancia de patrones (pattern matching), lo que facilita trabajar con estructuras complejas y mejora la legibilidad del código.

##### Ejemplo básico simulando switch de C++

Este ejemplo muestra cómo usar `match-case` para simular un `switch` tradicional, donde se compara un valor exacto contra casos específicos.

```python
def switch_example(value):
    match value:
        case 1:
            print("Uno")
        case 2:
            print("Dos")
        case 3:
            print("Tres")
        case _:
            print("Otro número")

switch_example(1)
switch_example(2)
switch_example(5)
```

##### Caso más complejo: patrones con tipos y condiciones

El `match-case` permite patrones más avanzados, como verificar tipos de datos y agregar condiciones con `if`.

```python
def process_data(data):
    match data:
        case int() as x if x > 0:
            print(f"Entero positivo: {x}")
        case int() as x if x < 0:
            print(f"Entero negativo: {x}")
        case str() as s:
            print(f"Cadena: {s.upper()}")
        case list() as l:
            print(f"Lista con {len(l)} elementos")
        case _:
            print("Tipo no reconocido")

process_data(42)
process_data(-5)
process_data("hola")
process_data([1, 2, 3])
```

##### Potencia del match-case: desestructuración y patrones complejos

Una de las fortalezas del `match-case` es la capacidad de desestructurar datos complejos, como tuplas, listas o diccionarios, y aplicar patrones específicos.

```python
def analyze_point(point):
    match point:
        case (0, 0):
            print("Origen")
        case (x, 0):
            print(f"Eje X en {x}")
        case (0, y):
            print(f"Eje Y en {y}")
        case (x, y) if x == y:
            print(f"Diagonal: {x}, {y}")
        case (x, y):
            print(f"Punto: {x}, {y}")

analyze_point((0, 0))
analyze_point((5, 0))
analyze_point((0, 3))
analyze_point((2, 2))
analyze_point((1, 3))
```
