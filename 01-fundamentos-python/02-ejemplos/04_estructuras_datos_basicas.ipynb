## 1. Listas

```python
frutas = ["manzana", "banana", "naranja"]
print("Primera fruta:", frutas[0])
frutas.append("kiwi")
print("Lista actualizada:", frutas)
```

```python
frutas2 = ["pera", "uva"]
frutas.extend(frutas2)
print("Lista extendida:", frutas)

frutas.insert(1, "mango")
print("Lista con mango insertado:", frutas) 
frutas.append("banana")
print("Lista con banana añadida:", frutas)

#ordenar la lista por longitud de nombre de fruta
frutas.sort(key=lambda x: len(x))
print("Lista ordenada:", frutas)
#eliminar todas las bananas
while "banana" in frutas:
    frutas.remove("banana")
print("Lista sin bananas:", frutas)
frutas.pop(2)
print("Lista después de eliminar el tercer elemento:", frutas)
print("Número de frutas en la lista:", len(frutas))
if "banana" not in frutas:
    print("No hay bananas en la lista.")


# listar las frutas con un bucle por indice
print("Frutas en la lista:")
for i in range(len(frutas)):
    print(f"{i}- {frutas[i]}")

del frutas[1]
print("Lista después de eliminar la segunda fruta:", frutas)


```

```python
#añadir mas frutas
frutas += ["fresa", "melocotón", "cereza","kiwi"]
print("Lista con más frutas añadidas:", frutas)
```

```python
# slicing
print("Primeras dos frutas:", frutas[:2])
print("Frutas desde la segunda en adelante:", frutas[1:])
print("Última fruta:", frutas[-1])
print("Frutas en orden inverso:", frutas[::-1])
print("Frutas desde la segunda hasta la penúltima:", frutas[1:-1])
print("Frutas desde la segunda hasta la cuarta:", frutas[1:4])
print("Frutas en posiciones impares:", frutas[::2])
print("Frutas en posiciones pares:", frutas[1::2])
print("Frutas avanzando de 2 en 2 desde la posición 1:", frutas[1::3])
print("Frutas desde la posición 1 hasta la 5 avanzando de 2 en 2:", frutas[1:5:2])
# posición de la fruta "kiwi"
print("Posición de la fruta 'kiwi':", frutas.index("kiwi") if "kiwi" in frutas else "No está en la lista")  
# contar cuantas veces aparece "fresa"
print("Número de veces que aparece 'fresa':", frutas.count("fresa"))

#concatenar listas
mas_frutas = ["papaya", "mango"]
todas_frutas = frutas + mas_frutas
print("Lista concatenada de todas las frutas:", todas_frutas)
#enumerar frutas
print("Frutas enumeradas:")
for idx, fruta in enumerate(frutas):
    print(f"{idx}: {fruta}")

#sumar los elementos de una lista
superfruta=frutas[1]+frutas[3]
print("Superfruta:", superfruta)

#map para convertir a mayusculas
frutas_mayusculas = list(map(str.upper, frutas))
print("Frutas en mayúsculas:", frutas_mayusculas)

#alias de lista
otras_frutas = frutas
otras_frutas.append("piña")
print("Lista original después de modificar el alias:", frutas)
print("Alias de la lista:", otras_frutas)
#funciones que de listas que devuelven listas vs funciones que modifican la lista original
frutas_copia = frutas.copy()  # crea una copia de la lista
frutas_copia.sort()  # ordena la copia
print("Lista original después de ordenar la copia:", frutas)
print("Copia ordenada de la lista:", frutas_copia)
sorted_frutas = sorted(frutas)  # devuelve una nueva lista ordenada
```

```python
#listas por comprensión
frutas_con_a = [fruta for fruta in frutas if 'a' in fruta]
print("Frutas que contienen la letra 'a':", frutas_con_a)
frutas_longitud = [len(fruta) for fruta in frutas]
print("Longitudes de los nombres de las frutas:", frutas_longitud)
frutas_mayusculas_comp = [fruta.upper() for fruta in frutas]
print("Frutas en mayúsculas (compresión):", frutas_mayusculas_comp)
#eliminar duplicados usando set
frutas_sin_duplicados = list(set(frutas))
print("Frutas sin duplicados:", frutas_sin_duplicados)
#obtener el cuadrado de todos los números en una lista
numeros = [1, 2, 3, 4, 5]
cuadrados = [x**2 for x in numeros]
print("Cuadrados de los números:", cuadrados)
```

## 2. Tuplas

```python
coordenadas = (10, 20)
tuplas1Elemento = (5,)
print("Coordenadas:", coordenadas)
print("Tupla de un elemento:", tuplas1Elemento)
print("X:", coordenadas[0])
print("Y:", coordenadas[1])
#cosa que no se puede hacer con tuplas
#coordenadas[0] = 15  # Esto daría un error
coordenadas=(coordenadas[0] + 5, coordenadas[1])  # Crear una nueva tupla
print("Coordenadas actualizadas:", coordenadas)
#otras cosas que no se pueden hacer con tuplas
#coordenadas.append(30)  # Esto daría un error
#coordenadas.remove(10)  # Esto daría un error
#coordenadas.pop()  # Esto daría un error

#desempaquetado de tuplas
x,y = coordenadas
x,y=coordenadas[0],coordenadas[1]

x,y=y,x

print("X desempaquetado:", x)
print("Y desempaquetado:", y)
#mas operaciones con tuplas
tupla_concatenada = coordenadas + (30, 40)
print("Tupla concatenada:", tupla_concatenada)
print("Longitud de la tupla concatenada:", len(tupla_concatenada))
print("Número de veces que aparece 10 en la tupla concatenada:", tupla_concatenada.count(10))
print("Índice de 20 en la tupla concatenada:", tupla_concatenada.index(20) if 20 in tupla_concatenada else "No está en la tupla")

#slicing en tuplas
print("Primeros dos elementos de la tupla concatenada:", tupla_concatenada[:2])
print("Elementos desde el segundo en adelante:", tupla_concatenada[1:])
print("Último elemento de la tupla concatenada:", tupla_concatenada[-1])
print("Tupla concatenada en orden inverso:", tupla_concatenada[::-1])
print("Elementos desde el segundo hasta el penúltimo:", tupla_concatenada[1:-1])
print("Elementos desde el segundo hasta el cuarto:", tupla_concatenada[1:4])
print("Elementos en posiciones impares:", tupla_concatenada[::2])
print("Elementos en posiciones pares:", tupla_concatenada[1::2])
```

## 3. Diccionarios

```python
persona = {"nombre": "Laura", "edad": 25}
print("Nombre:", persona["nombre"])
persona["edad"] += 1
print("Edad actualizada:", persona["edad"])
```

```python
# Creación de diccionarios
# Literal
d1 = {'nombre': 'Ana', 'edad': 30}

# constructor dict()
d2 = dict(ciudad='Madrid', pais='España')
# desde secuencia de pares
d3 = dict([('x', 10), ('y', 20)])
# fromkeys para crear con valores por defecto
keys = ['a', 'b', 'c']
d4 = dict.fromkeys(keys, 0)
# diccionario vacío
dv = {} # diccionario vacío
dvacio = dict() # diccionario vacío
print('d1 =', d1)
print('d2 =', d2)
print('d3 =', d3)
print('d4 =', d4)
print('dv =', dv)
```

```python
notas = {'Ana':8, 'Luis':7.5, 'Marta':9}
print("Nota de Ana:", notas['Ana'])
notas['Luis'] = 8.0
print("Nota actualizada de Luis:", notas['Luis'])
notas['Carlos'] = 7.0
print("Notas después de añadir a Carlos:", notas)

print(notas.keys())
print(notas.values())
print(notas.items())
```

```python
# Manejo de errores al acceder a claves inexistentes
# notas['Juan']
print(notas.get('Juan', 'No está en la lista'))
print(notas.get('Ana', 'No está en la lista'))
# Eliminar elementos
print(notas.pop('Carlos', None) ) # Elimina a Carlos si existe
print("Notas después de eliminar a Carlos:", notas)
print(notas.popitem())  # Elimina un par arbitrario
```

```python
# clear
notas.clear()
print("Notas después de clear:", notas)
```

```python
# Actualizar con otro diccionario
notas2 = {'Eva': 9.5, 'Luis': 8.5}
notas.update(notas2)
print("Notas después de update con notas2:", notas) 

# Está en el diccionario?
print("¿Está Ana en el diccionario?", 'Ana' in notas)
print("¿Está Juan en el diccionario?", 'Juan' in notas)

# longitud del diccionario
print("Número de elementos en el diccionario:", len(notas))

# copiar diccionario
notas_copia = notas.copy()
print("Copia del diccionario notas:", notas_copia)
```

```python
for alumno, nota in notas.items():
    print(f"{alumno}: {nota}")

for alumno in notas.keys():
    print(f"Alumno: {alumno}")

for nota in notas.values():
    print(f"Nota: {nota}") 

# comprensión de diccionarios
notas['Ana'] = 6
notas_altas = {alumno: nota for alumno, nota in notas.items() if nota >= 8}
print("Diccionario de notas altas:", notas_altas)

# diccionario con números y sus cuadrados
numero_cuadrado = {x: x**2 for x in range(6)}
print("Diccionario de números y sus cuadrados:", numero_cuadrado)
# eliminar duplicados de una lista usando diccionario
lista_con_duplicados = ['manzana', 'banana', 'manzana', 'kiwi', 'banana']
sin_duplicados = list(dict.fromkeys(lista_con_duplicados))
print("Lista sin duplicados usando diccionario:", sin_duplicados)

# convertir dos listas en un diccionario
```

## 4. Conjuntos

```python
colores = {"rojo", "verde", "azul"}
colores.add("amarillo")
print("Colores:", colores)
colores.add("rojo")  # No añade duplicados
print("Colores después de intentar añadir 'rojo' de nuevo:", colores)

# eliminar elementos
colores.discard("cian")  # Elimina cian si existe
print("Colores después de eliminar 'verde':", colores)
colores.remove("verde")  # Elimina verde, da error si no existe
print("Colores después de eliminar 'verde':", colores)

# comprobrar si un elemento está en el conjunto
print("¿Está 'azul' en el conjunto?", 'azul' in colores)
print("¿Está 'verde' en el conjunto?", 'verde' in colores)


```

```python
#extraer un elemento arbitrario y eliminarlo
elemento = colores.pop()
print("Elemento extraído y eliminado:", elemento)
print("Conjunto después de pop:", colores)
```

```python
# operaciones con conjuntos
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
print("Conjunto A:", A)
print("Conjunto B:", B)
print("Unión de A y B:", A | B)
print("Intersección de A y B:", A & B)
print("Diferencia de A y B (A - B):", A - B)
print("Diferencia de B y A (B - A):", B - A)
print("Diferencia simétrica de A y B:", A ^ B)
```

```python
# vaciar conjunto
colores.clear()
print("Conjunto después de clear:", colores)
```
