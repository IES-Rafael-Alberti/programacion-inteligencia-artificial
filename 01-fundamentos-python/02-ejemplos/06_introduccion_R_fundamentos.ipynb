# Introducción a R: Fundamentos para Inteligencia Artificial

Este notebook te introduce a los conceptos fundamentales de R, un lenguaje utilizado frecuentemente en ciencia de datos, estadística y análisis.

Veremos:
- Variables y tipos de datos
- Operadores
- Estructuras de control (`if`, `for`, `while`)
- Funciones
- Estructuras de datos básicas (vectores, listas, data frames)

---

## 1. Variables y tipos de datos

```
nombre <- "Ana"
edad <- 25
altura <- 1.65
es_estudiante <- TRUE

print(nombre)
print(edad)
print(altura)
print(es_estudiante)
```

## 2. Operadores aritméticos y lógicos

```
a <- 10
b <- 3

a + b   # Suma
a - b   # Resta
a * b   # Multiplicación
a / b   # División
a %% b  # Módulo
a ^ b   # Potencia

a > b   # Mayor que
a == b  # Igualdad
a != b  # Distinto
```

## 3. Control de flujo

```
if (edad >= 18) {
  print("Mayor de edad")
} else {
  print("Menor de edad")
}

# Bucle for
for (i in 1:5) {
  print(i)
}

# Bucle while
i <- 1
while (i <= 5) {
  print(i)
  i <- i + 1
}
```

## 4. Funciones

```
saludar <- function(nombre) {
  print(paste("Hola", nombre))
}

saludar("Luis")

cuadrado <- function(x) {
  return(x^2)
}

cuadrado(4)
```

## 5. Estructuras de datos básicas

```
# Vectores
notas <- c(7, 8, 6)
print(notas)

# Listas
persona <- list(nombre="Carlos", edad=30, aprobado=TRUE)
print(persona)

# Data Frame
alumnos <- data.frame(
  nombre = c("Ana", "Luis", "Marta"),
  nota = c(7, 4, 9)
)
print(alumnos)

# Acceso
alumnos$nota[alumnos$nombre == "Luis"]
```
