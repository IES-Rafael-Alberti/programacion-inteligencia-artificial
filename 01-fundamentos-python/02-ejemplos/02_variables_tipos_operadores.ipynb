# 02 - Variables, tipos de datos y operadores en Python

En este notebook aprenderás a:

- Declarar y usar variables en Python.
- Identificar los principales tipos de datos: enteros, flotantes, cadenas, booleanos.
- Usar operadores aritméticos, relacionales y lógicos.

---

## 1. Variables en Python

```python
# Declaración de variables
nombre = "Juan"
edad = 20
es_estudiante = range(1, 10, 2)

print(f"Nombre: {nombre}. El tipo de la variable es {type(nombre)}")
print(f"Edad: {edad}. El tipo de la variable es {type(edad)}")
print(f"¿Es estudiante? {es_estudiante}. El tipo de la variable es {type(es_estudiante)}")
```

## 2. Tipos de datos

```python
print(type(nombre))       # str
print(type(edad))         # int
print(type(es_estudiante)) # bool
print(type(3.1415))       # float
```

## 3. Operadores aritméticos

```python
a = 10
b = 3

print("Suma:", a + b)
print("Resta:", a - b)
print("Multiplicación:", a * b)
print("División:", a / float(b))
print("División entera:", a // b)
print("Módulo:", a % b)
print("Potencia:", a ** b)
```

## 4. Operadores relacionales

```python
print("¿a es mayor que b?", a > b)
print("¿a es igual a b?", a == b)
print("¿a es distinto de b?", a != b)
```

## 5. Operadores lógicos

```python
x = True
y = False

print(f"{x} and {y}: {x and y}")
print(f"{y} or {x}: {x or y}")
print(f"not {x}: {not x}")
```

## Operadores binarios

```python
x = 1
y = 0

print("x & y",) x & y  # AND bit a bit
print("x | y", x | y)  # OR bit a bit
print("x ^ y", x ^ y)  # XOR bit a bit
print("~x", ~x)        # NOT bit a bit
print("x << 1", x << 1) # Desplazamiento a la izquierda
print("x >> 1", x >> 1) # Desplazamiento a la derecha
```
