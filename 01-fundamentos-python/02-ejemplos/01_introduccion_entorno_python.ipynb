## 2. Formato en Markdown

Puedes escribir títulos, listas, texto en **negrita**, *cursiva*, `código`, enlaces y más.

Ejemplos:
- `#` Título de nivel 1
- `##` Título de nivel 2
- `*texto*` cursiva
- `**texto**` negrita
- `` `código` `` para resaltar código
- `[enlace](https://www.ejemplo.com)` para enlaces
- Listas con `-` o `*`
- Listas numeradas con `1.`, `2.`, etc.

Puedes combinar estos formatos para crear explicaciones claras y bien estructuradas. ¡Practica escribiendo tus propias celdas Markdown!
