## 🧾 5. RESUMEN Y COMPARACIÓN

| Tipo      | Mutable | Ordenado | Permite duplicados | Sintaxis principal | Ejemplo          |
| --------- | ------- | -------- | ------------------ | ------------------ | ---------------- |
| **list**  | ✅ Sí    | ✅ Sí     | ✅ Sí               | `[ ]`              | `[1, 2, 3]`      |
| **tuple** | ❌ No    | ✅ Sí     | ✅ Sí               | `( )`              | `(1, 2, 3)`      |
| **dict**  | ✅ Sí    | ✅ (3.7+) | ❌ Claves únicas    | `{clave: valor}`   | `{'a':1, 'b':2}` |
| **set**   | ✅ Sí    | ❌ No     | ❌ No               | `{ }` o `set()`    | `{1, 2, 3}`      |
