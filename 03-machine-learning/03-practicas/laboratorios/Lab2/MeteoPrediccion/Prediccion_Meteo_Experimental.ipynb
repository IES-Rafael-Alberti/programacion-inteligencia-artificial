# 🌦️ Predicción Meteorológica – Notebook Experimental

Este cuaderno guía la exploración y modelado de datos meteorológicos usando un dataset previamente construido (por ejemplo, Open-Meteo + AEMET). El objetivo es predecir una variable como la temperatura o humedad usando otras variables del conjunto de datos.

## 📥 1. Carga del dataset

```
import pandas as pd

df = pd.read_csv("dataset_meteo.csv", parse_dates=["time"])
df.set_index("time", inplace=True)
df.head()
```

## 🔍 2. Análisis exploratorio

Objetivo: Comprender la evolución y relaciones entre variables para seleccionar predictores útiles.

```
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 4))
sns.lineplot(data=df[['temperature_2m', 'relative_humidity_2m']])
plt.title("Temperatura y humedad a lo largo del tiempo")
plt.grid()
plt.tight_layout()
plt.show()
```

```
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Matriz de correlación")
plt.show()
```

## 🧹 3. Preparación de datos

Objetivo: crear un conjunto limpio y entrenable.

```
df = df.copy()
df['hour'] = df.index.hour
df['dayofweek'] = df.index.dayofweek

# Variable objetivo (ejemplo: temperatura)
y = df['temperature_2m']

# Features iniciales
X1 = df[['relative_humidity_2m', 'wind_speed_10m']]
X2 = df[['relative_humidity_2m', 'wind_speed_10m', 'hour', 'dayofweek']]
```

## 🤖 4. Entrenamiento de modelos

Prueba con cualquier modelo: scikit-learn, cuML, XGBoost, etc.

```
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

X_train, X_test, y_train, y_test = train_test_split(X1, y, test_size=0.2, random_state=42)

model = RandomForestRegressor()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mse1 = mean_squared_error(y_test, y_pred)
print(f"MSE con 2 features: {mse1:.2f}")
```

```
# Con más features
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y, test_size=0.2, random_state=42)

model2 = RandomForestRegressor()
model2.fit(X_train2, y_train2)
y_pred2 = model2.predict(X_test2)

mse2 = mean_squared_error(y_test2, y_pred2)
print(f"MSE con más features: {mse2:.2f}")
```
