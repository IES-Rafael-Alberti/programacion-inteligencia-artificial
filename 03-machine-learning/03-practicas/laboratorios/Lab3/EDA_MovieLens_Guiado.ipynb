# 🎬 EDA Guiado – Dataset MovieLens

Este notebook realiza un Análisis Exploratorio de Datos completo sobre un subconjunto del dataset **MovieLens**. Incluye carga de datos, análisis estadístico, visualización, detección de valores atípicos y conclusiones orientadas a selección de modelo.

## 📦 1. Carga de datos

````
import pandas as pd

# Cargar los datasets
ratings = pd.read_csv('https://files.grouplens.org/datasets/movielens/ml-latest-small/ratings.csv')
movies = pd.read_csv('https://files.grouplens.org/datasets/movielens/ml-latest-small/movies.csv')

# Unir ambos
df = pd.merge(ratings, movies, on='movieId')
df.head()
````

## 🧼 2. Revisión y limpieza

````
df.info()
df.isnull().sum()
df.describe()
````

## 📊 3. Análisis de frecuencia y distribución de ratings

````
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,4))
sns.countplot(x='rating', data=df)
plt.title('Distribución de valoraciones')
plt.xlabel('Puntuación')
plt.ylabel('Frecuencia')
plt.show()
````

## 📈 4. Películas con más valoraciones

````
df['title'].value_counts().head(10)
````

## ⏳ 5. Evolución temporal de las valoraciones

````
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')
df['year'] = df['timestamp'].dt.year

plt.figure(figsize=(10,5))
df.groupby('year')['rating'].count().plot()
plt.title('Cantidad de valoraciones por año')
plt.xlabel('Año')
plt.ylabel('Nº valoraciones')
plt.grid(True)
plt.show()
````

## 🧪 6. ¿Qué podríamos predecir?

- **Clasificación:** ¿Le dará un usuario una buena valoración (>=4)? (binario)
- **Regresión:** ¿Qué nota le pondrá un usuario a una película específica?
- **Ranking/Recomendación:** ¿Qué películas recomendar al usuario basándonos en otras que le han gustado?

## ✅ 7. Conclusiones y preparación para modelado

- La variable `rating` está bastante centrada en torno a 3.5–4.0, con pocas puntuaciones extremas.
- Algunas películas tienen muchas más valoraciones que otras (popularidad).
- El dataset tiene valores limpios y no contiene nulos.
- Para tareas iniciales de clasificación binaria, podemos transformar la columna `rating` en una nueva columna `target`:

```python
df['target'] = (df['rating'] >= 4.0).astype(int)
```
