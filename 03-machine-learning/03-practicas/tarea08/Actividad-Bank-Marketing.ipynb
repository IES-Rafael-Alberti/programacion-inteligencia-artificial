```python
# Importa las librerías necesarias
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import joblib
```

## 1. Carga del dataset
Explica aquí dónde descargaste el archivo y los pasos previos.

```python
csv_path = 'datos/bank-additional-full.csv'  # Ajusta la ruta si es necesario
df = pd.read_csv(csv_path, sep=';')
df.head()
```

## 2. EDA breve
Describe la distribución de la variable objetivo, las categorías importantes y posibles valores atípicos.

```python
df['y'].value_counts(normalize=True)
```

## 3. Limpieza
Documenta las decisiones de limpieza (columnas eliminadas, conversión de la variable objetivo, tratamiento de nulos).

```python
df = df.drop(columns=['duration'])  # Justifica la eliminación si la decisión aplica
df['y'] = df['y'].map({'yes': 1, 'no': 0})
df.isnull().sum()
```

## 4. Codificación de variables categóricas
Identifica las columnas categóricas y aplica One-Hot Encoding cuando proceda.

```python
categorical_columns = df.select_dtypes(include='object').columns
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
encoded = encoder.fit_transform(df[categorical_columns])
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(categorical_columns))
df_encoded = pd.concat([df.drop(columns=categorical_columns).reset_index(drop=True), encoded_df], axis=1)
```

## 5. Escalado de variables numéricas
Decide qué columnas escalarás y por qué. Describe el impacto esperado.

```python
numeric_columns = ['age', 'balance', 'campaign', 'previous']  # Ajusta según tu análisis
scaler = StandardScaler()
df_encoded[numeric_columns] = scaler.fit_transform(df_encoded[numeric_columns])
```

## 6. División train/test
Indica si estratificas y por qué.

```python
X = df_encoded.drop(columns='y')
y = df_encoded['y']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

## 7. Guardar conjuntos preparados
Guarda en formatos gestionables y explica por qué elegiste cada uno.

```python
joblib.dump(X_train, 'Tarea/X_train.joblib')
joblib.dump(X_test, 'Tarea/X_test.joblib')
joblib.dump(y_train, 'Tarea/y_train.joblib')
joblib.dump(y_test, 'Tarea/y_test.joblib')
```
