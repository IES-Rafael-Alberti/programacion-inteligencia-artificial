## 🧑‍💻 Carga de datos y exploración inicial

````
import pandas as pd
from pycaret.datasets import get_data

# Cargar datos
data = get_data('diamond')  # Alternativa más ligera si Ames falla
data.head()
````

## ⚙️ Configuración del entorno de trabajo con `setup`
PyCaret necesita una configuración inicial donde:
- Se define la variable objetivo (`target`)
- Se detectan y transforman automáticamente las columnas
- Se activa el escalado, codificación y detección de outliers

````
from pycaret.regression import *

reg = setup(
    data=data,                         # DataFrame de entrada con variables y target
    target='Price',                    # Variable objetivo a predecir
    normalize=True,                    # Escalado de variables numéricas (útil para modelos sensibles a escala)
    transformation=True,               # Transformación de la variable objetivo (mejora normalidad/heterocedasticidad)
    feature_selection=True,            # Selección automática de variables más relevantes
    remove_multicollinearity=True,     # Elimina variables altamente correlacionadas
    multicollinearity_threshold=0.9,   # Umbral de correlación para eliminar colinealidad
    session_id=123,                    # Semilla para reproducibilidad
    silent=True                        # Ejecuta setup sin prompts interactivos
)
````

## 🧩 Opciones de `setup` más usadas y relevantes (clasificación y regresión)

A continuación, parámetros **frecuentes** en PyCaret (válidos para `regression` y `classification`, cuando aplique):

### 🔧 Control del experimento
- **`session_id`**: fija la semilla para reproducibilidad.
- **`train_size`**: proporción para entrenamiento (p. ej. `0.7`).
- **`fold` / `fold_strategy`**: número de folds y estrategia de validación (`kfold`, `stratifiedkfold`, etc.).

### 🧹 Limpieza y preparación de datos
- **`ignore_features`**: lista de columnas a excluir (IDs, campos irrelevantes).
- **`numeric_features` / `categorical_features`**: fuerza el tipo de columnas si la detección automática falla.
- **`date_features`**: columnas de fecha para ingeniería automática.
- **`imputation_type`**: estrategia de imputación (`simple`, `iterative`, etc.).
- **`normalize`**: escalado de variables numéricas.
- **`transformation`**: transforma variables numéricas para mejorar distribuciones.
- **`remove_outliers`**: elimina outliers detectados automáticamente.
- **`outliers_threshold`**: sensibilidad de detección de outliers.
- **`remove_multicollinearity`**: elimina variables con alta correlación.
- **`multicollinearity_threshold`**: umbral de colinealidad.

### 🧠 Ingeniería de variables
- **`feature_selection`**: activa selección automática de features.
- **`feature_selection_method`**: método de selección (p. ej. `classic`, `univariate`).
- **`polynomial_features`**: genera combinaciones polinómicas (útil en regresión).
- **`bin_numeric_features`**: discretiza variables numéricas en bins.
- **`combine_rare_levels`**: agrupa categorías poco frecuentes.
- **`rare_level_threshold`**: porcentaje mínimo para no considerarse “raro”.

### ⚖️ Específico de clasificación
- **`fix_imbalance`**: aplica técnicas para clases desbalanceadas (SMOTE, etc.).
- **`fix_imbalance_method`**: método de balanceo.
- **`target`** (en clasificación): variable objetivo (etiqueta de clase).

### ✅ Consejos rápidos
- **Clasificación**: prioriza `fix_imbalance`, `fold_strategy='stratifiedkfold'`.
- **Regresión**: revisa `transformation`, `remove_outliers`, `normalize` y `polynomial_features`.

> Tip: `setup` genera un **pipeline completo** que luego se aplica automáticamente a cualquier modelo entrenado con PyCaret.
