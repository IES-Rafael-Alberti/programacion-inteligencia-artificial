# 🌐 Hopsworks – Uso de Feature Store con datos meteorológicos

Este notebook muestra cómo subir un dataset meteorológico a Hopsworks y cómo leerlo posteriormente desde otro entorno para usarlo en entrenamiento de modelos.

## 🧩 Requisitos previos
- Tener una cuenta en [Hopsworks](https://www.hopsworks.ai/signup)
- Crear un proyecto en su panel web
- Instalar la librería:
```bash
pip install hopsworks
```

## 🔑 1. Conexión con Hopsworks

````
import hopsworks

project = hopsworks.login()
fs = project.get_feature_store()
````

## 📤 2. Subida del dataset como Feature Group

````
import pandas as pd

df = pd.read_csv("dataset_meteo.csv", parse_dates=["time"])

# Crear feature group
fg = fs.get_or_create_feature_group(
    name="meteo_sevilla",
    version=1,
    description="Datos meteo Open-Meteo + AEMET para Sevilla",
    primary_key=["time"],
    event_time="time"
)

fg.insert(df, write_options={"wait_for_job": True})
````

## 📥 3. Lectura del Feature Group desde otro entorno o notebook

````
# Repetir login si es otro entorno
project = hopsworks.login()
fs = project.get_feature_store()

# Leer dataset
fg = fs.get_feature_group("meteo_sevilla", version=1)
df = fg.read()
df.head()
````
