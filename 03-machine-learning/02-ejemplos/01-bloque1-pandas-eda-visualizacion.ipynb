Este cuaderno está pensado para practicar de forma intensiva el manejo de `pandas`, la exploración de datos (EDA) y técnicas básicas de visualización. Al final encontrarás enlaces a herramientas alternativas como `polars` y `cuDF`.

Dataset usado: `weather.csv` (puedes sustituirlo por otro si lo prefieres).

## 1. Carga del dataset y primeras inspecciones

````
import pandas as pd

df = pd.read_csv("weather.csv", parse_dates=["date"])
df.head()
````

````
df.info()
````

````
df.describe(numeric_only=True)
````

````
df.isnull().sum()
````

## 2. Limpieza de datos

````
# Eliminar duplicados
df.drop_duplicates(inplace=True)

# Eliminar columnas con muchos nulos o irrelevantes
df = df.drop(columns=["station_id", "remarks"], errors="ignore")

# Rellenar valores faltantes
df['temperature'] = df['temperature'].fillna(method='ffill')
````

## 3. Transformaciones útiles

````
# Crear columnas de fecha
df['month'] = df['date'].dt.month
df['dayofweek'] = df['date'].dt.dayofweek

# Agrupar por mes y calcular medias
df_monthly = df.groupby('month')[['temperature', 'humidity']].mean()
df_monthly
````

## 4. Visualización de datos

Usamos seaborn, matplotlib y plotly para ver patrones.

````
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='date', y='temperature')
plt.title("Temperatura diaria")
plt.grid()
````

````
sns.boxplot(data=df, x='month', y='temperature')
plt.title("Distribución mensual de temperaturas")
````

````
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
````

````
import plotly.express as px

px.scatter(df, x='humidity', y='temperature', color='month', title="Relación humedad-temperatura")
````

## 🧭 Alternativas modernas a pandas

### 🔸 [Polars (documentación)](https://pola-rs.github.io/polars-book/)
```python
import polars as pl
pl.read_csv("weather.csv").groupby("month").mean()
```

### 🔸 [cuDF (documentación)](https://docs.rapids.ai/api/cudf/stable/)
```python
import cudf
df = cudf.read_csv("weather.csv")
df.groupby("month").mean()
```

> Recuerda que Polars es mucho más rápido en CPU, y cuDF en GPU, pero tienen menor ecosistema que `pandas`. Son buenas opciones para trabajar con grandes volúmenes o notebooks pesados.
