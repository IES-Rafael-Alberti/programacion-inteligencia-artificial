# Dataset de Clima con Open-Meteo

Este notebook descarga datos meteorológicos desde la API pública de Open-Meteo para una ciudad específica, los procesa y los guarda en formatos CSV y Parquet.

```
# !pip install requests pandas pyarrow
```

```
import requests
import pandas as pd
```

## Configuración de la consulta

```
lat, lon = 37.39, -5.99  # Sevilla
url = f"https://api.open-meteo.com/v1/forecast?latitude={lat}&longitude={lon}&hourly=temperature_2m,relative_humidity_2m&timezone=auto"
response = requests.get(url)
data = response.json()
```

## Conversión a DataFrame

```
df = pd.DataFrame(data['hourly'])
df['time'] = pd.to_datetime(df['time'])
df.set_index('time', inplace=True)
df.head()
```

## Guardar en CSV y Parquet

```
df.to_csv('data/clima_sevilla.csv')
df.to_parquet('data/clima_sevilla.parquet')
```

Este conjunto de datos puede utilizarse como base para análisis de series temporales o para servir desde una API con FastAPI.
