# API con FastAPI para Servir Datos Climáticos

Este notebook crea un pequeño servidor con FastAPI que expone un endpoint para servir un subconjunto del dataset `clima_sevilla.csv`.

```
# !pip install fastapi uvicorn pandas
```

```
from fastapi import FastAPI
import pandas as pd
from typing import List
from pydantic import BaseModel
```

## Inicialización de la API

```
app = FastAPI()
df = pd.read_csv('data/clima_sevilla.csv')
```

## Endpoint para devolver datos aleatorios

```
@app.get("/datos")
def get_sample(n: int = 5):
    return df.sample(n).to_dict(orient="records")
```

## Cómo ejecutarlo desde terminal (si estuviera fuera del notebook)

```
uvicorn nombre_modulo:app --reload
```

Este servidor puede integrarse dentro del proyecto de Poetry o como microservicio para servir datos a notebooks, dashboards o APIs externas.
