# 103 - Plantilla de Dashboard (Gradio/Streamlit) — Base

Elige **Gradio** o **Streamlit** para un dashboard simple.
Por defecto ponemos un ejemplo con **Gradio**.

```
import json
try:
    import gradio as gr
    HAS_GRADIO = True
except Exception:
    HAS_GRADIO = False

def predictor(f1: float, f2: float) -> float:
    # Sustituir por carga real del modelo y predicción
    return 1.0 if (f1 + f2) > 0 else 0.0

if HAS_GRADIO:
    demo = gr.Interface(
        fn=predictor,
        inputs=[gr.Slider(-3,3,0,label='feature1'), gr.Slider(-3,3,0,label='feature2')],
        outputs=gr.Number(label='Predicción'),
        title='Demo de Predicción',
        description='Sustituye por tu modelo entrenado.'
    )
    print('Ejecuta demo.launch() para abrir el dashboard.')
else:
    print('Instala gradio o adapta a Streamlit.')
```
