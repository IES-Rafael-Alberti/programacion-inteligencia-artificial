# 104 - Plantilla de Pipeline (Prefect/Airflow) — Base

Ejemplo mínimo de **pipeline ETL** con **Prefect**. Si no está disponible, se muestra versión Python pura.

```
try:
    from prefect import flow, task
    HAS_PREFECT = True
except Exception:
    HAS_PREFECT = False

if HAS_PREFECT:
    @task
    def extract(path='data/dataset.csv'):
        import pandas as pd
        try:
            return pd.read_csv(path)
        except Exception:
            return None

    @task
    def transform(df):
        if df is None:
            return None
        return df.dropna()

    @task
    def load(df, path='artifacts/clean.csv'):
        if df is None:
            return False
        from pathlib import Path
        Path('artifacts').mkdir(exist_ok=True)
        df.to_csv(path, index=False)
        return True

    @flow
    def etl():
        d = extract()
        t = transform(d)
        ok = load(t)
        return ok

    print('Ejecuta etl() para lanzar el pipeline con Prefect.')
else:
    # Versión Python pura
    import pandas as pd
    from pathlib import Path
    def etl(path_in='data/dataset.csv', path_out='artifacts/clean.csv'):
        try:
            df = pd.read_csv(path_in)
        except Exception:
            df = pd.DataFrame({'feature1':[0,1], 'feature2':[1,-1], 'target':[1,0]})
        df = df.dropna()
        Path('artifacts').mkdir(exist_ok=True)
        df.to_csv(path_out, index=False)
        return True
    print('Ejecuta etl() (versión Python pura).')
```
