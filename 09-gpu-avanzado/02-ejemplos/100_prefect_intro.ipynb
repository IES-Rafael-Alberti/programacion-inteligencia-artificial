# 100 - Prefect para orquestación de flujos — Base

Introducción a **Prefect**, herramienta para definir y ejecutar flujos de trabajo.
Simularemos un **ETL sencillo**.

```
try:
    from prefect import flow, task
    HAS_PREFECT = True
except Exception as e:
    HAS_PREFECT = False; print('⚠️ Prefect no disponible:', e)

if HAS_PREFECT:
    @task
    def extract():
        return [1,2,3,4,5]

    @task
    def transform(data):
        return [x*2 for x in data]

    @task
    def load(data):
        print('Datos cargados:', data)

    @flow
    def etl_flow():
        d = extract()
        t = transform(d)
        load(t)

    print('Ejecuta etl_flow() para probar el flujo.')
```
