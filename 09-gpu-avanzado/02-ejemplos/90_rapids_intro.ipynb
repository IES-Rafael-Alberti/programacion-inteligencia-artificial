# 90 - Introducción a RAPIDS (cuDF, cuML) — Base

RAPIDS es un ecosistema de NVIDIA que acelera ciencia de datos en **GPU**.

**Objetivos**
- Crear un DataFrame con **cuDF** (si está disponible).
- Entrenar un modelo simple con **cuML** (regresión logística) si el entorno lo permite.
- Mostrar alternativas y mensajes claros si no hay GPU.

```
try:
    import cudf
    HAS_CUDF = True
    print('✅ cuDF disponible')
except Exception as e:
    HAS_CUDF = False
    print('⚠️ cuDF no disponible o sin GPU:', e)

try:
    from cuml.linear_model import LogisticRegression as cuLogReg
    HAS_CUML = True
    print('✅ cuML disponible')
except Exception as e:
    HAS_CUML = False
    print('⚠️ cuML no disponible:', e)
```

```
# Demo mínima con cuDF (condicional)
if HAS_CUDF:
    gdf = cudf.DataFrame({'x1':[0.1,0.5,1.2,2.3],'x2':[1.0,0.3,0.7,1.1],'y':[0,0,1,1]})
    display(gdf)
else:
    print('Ejecuta en Colab RAPIDS o equipo con CUDA para cuDF.')
```

```
# Regresión logística con cuML (si está)
if HAS_CUML and HAS_CUDF:
    X = gdf[['x1','x2']].astype('float32')
    y = gdf['y'].astype('int32')
    clf = cuLogReg(max_iter=1000)
    clf.fit(X, y)
    preds = clf.predict(X)
    print('Predicciones:', preds.to_pandas().tolist())
else:
    print('Salta entrenamiento cuML por falta de entorno.')
```
