# 99 - Virtual Agent con LangChain + Gradio — Base

En este notebook construiremos un **agente conversacional simple** usando LangChain (simulado) y Gradio como interfaz.

```
try:
    import gradio as gr
    HAS_GRADIO = True
except Exception as e:
    HAS_GRADIO = False; print('⚠️ Falta Gradio:', e)

class DummyAgent:
    def __init__(self): self.history=[]
    def ask(self, query):
        self.history.append(query)
        return f'Respuesta simulada a: {query}'

agent = DummyAgent()
def chat_fn(msg):
    return agent.ask(msg)

if HAS_GRADIO:
    demo = gr.Interface(fn=chat_fn, inputs='text', outputs='text', title='Agente Virtual (demo)')
    print('Ejecuta demo.launch() para abrir la interfaz.')
```
