# 101 - Agente conectado a API externa — Base

Este ejemplo muestra cómo un **agente** puede consultar una **API externa** para enriquecer sus respuestas.

```
try:
    import requests
    HAS_REQ = True
except Exception as e:
    HAS_REQ = False; print('⚠️ requests no disponible:', e)

def ask_api(query):
    if not HAS_REQ:
        return '(simulado) No se pudo acceder a la API.'
    try:
        url = 'https://api.chucknorris.io/jokes/random'
        r = requests.get(url, timeout=5)
        if r.status_code==200:
            return r.json().get('value','(sin valor)')
        return '(fallback) API no respondió.'
    except Exception:
        return '(offline) Error de conexión.'

print('Ejemplo de respuesta de API:', ask_api('chiste'))
```
