# 91 - Introducción a JAX — Base

JAX ofrece **arrays como NumPy** con compilación en **GPU/TPU** y **autodiferenciación**.

**Objetivos**
- Crear arrays y operar con `jax.numpy`.
- Usar `jax.grad` para derivar funciones.
- Ejecutar en CPU si no hay GPU (funciona igual).

```
try:
    import jax
    import jax.numpy as jnp
    HAS_JAX = True
    print('✅ JAX', jax.__version__)
    print('Dispositivos:', jax.devices())
except Exception as e:
    HAS_JAX = False
    print('⚠️ JAX no disponible:', e)
```

```
if HAS_JAX:
    x = jnp.linspace(-3, 3, 7)
    f = lambda t: jnp.sin(t) * jnp.exp(-t**2)
    g = jax.grad(lambda t: jnp.sum(f(t)))
    print('x =', x)
    print('f(x) =', f(x))
    print('∂/∂x sum f(x) evaluado en x[0]:', float(g(x)[0]))
else:
    print('Instala jax + jaxlib para probar.')
```
