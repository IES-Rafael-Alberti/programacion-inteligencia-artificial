# 92 - Benchmarks CPU vs GPU (NumPy, cuDF/cuML, JAX) — Base

Compararemos tiempos de operaciones matriciales y de ML entre CPU y GPU cuando sea posible.

Incluye **checks condicionales** para que el notebook sea portable.

```
import numpy as np, time
N = 2000
A = np.random.rand(N, N).astype('float32')
t0 = time.time(); _ = A @ A; t_cpu = time.time()-t0
print('Prod. matricial NumPy (CPU):', round(t_cpu,3),'s')
```

```
# JAX (si está)
try:
    import jax, jax.numpy as jnp
    HAS_JAX=True
    B = jnp.array(A)
    t0=time.time(); _ = B @ B; t_jax = time.time()-t0
    print('Prod. matricial JAX (dispositivo actual):', round(t_jax,3),'s')
except Exception as e:
    HAS_JAX=False; print('JAX no disponible:', e)
```

```
# cuDF + cuML ejemplo mínimo (condicional)
try:
    import cudf
    from cuml.linear_model import LogisticRegression as cuLogReg
    HAS_RAPIDS=True
    import pandas as pd
    import numpy as np
    df = pd.DataFrame({'x1':np.random.rand(5000),'x2':np.random.rand(5000),'y':np.random.randint(0,2,5000)})
    gdf = cudf.DataFrame.from_pandas(df)
    X = gdf[['x1','x2']].astype('float32'); y=gdf['y'].astype('int32')
    from time import time as now
    t0=now(); cuLogReg(max_iter=200).fit(X,y); t_fit=now()-t0
    print('cuML LogisticRegression fit:', round(t_fit,3),'s')
except Exception as e:
    HAS_RAPIDS=False; print('RAPIDS no disponible:', e)
```
