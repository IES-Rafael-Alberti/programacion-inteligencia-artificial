TensorFlow Playground es una herramienta fantástica para visualizar y comprender cómo funcionan los modelos de redes neuronales, y sus datasets generados son muy útiles para ilustrar diferentes tipos de problemas de clasificación.

Afortunadamente, puedes generar datos similares a los que ves en TensorFlow Playground utilizando bibliotecas de Python como `scikit-learn` o incluso directamente con `NumPy`. Aquí te presento algunas formas de generar datasets similares a los que se encuentran en Playground, junto con ejemplos de código en Python:

**1. Datasets Lineales:**

Estos son los datasets más simples, donde las clases son linealmente separables.

* **En TensorFlow Playground:** Suele haber un dataset con dos nubes de puntos claramente separadas por una línea.

* **Generación con `scikit-learn`:** Puedes usar `make_classification` de `sklearn.datasets`.


```python
from sklearn.datasets import make_classification
from sklearn.datasets import make_circles, make_moons, make_blobs
import matplotlib.pyplot as plt
import numpy as np

# Generar un dataset lineal
X, y = make_classification(n_samples=200, n_features=2, n_informative=2, n_redundant=0, n_classes=2, n_clusters_per_class=1, random_state=42)



# Visualizar los datos
plt.figure(figsize=(6, 4))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdYlBu, edgecolors='k')
plt.title("Dataset Lineal")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

print("Forma de los datos (X, y):", X.shape, y.shape)
```


**2. Datasets No Lineales (Círculos):**

Estos datasets requieren una frontera de decisión no lineal para separar las clases.

* **En TensorFlow Playground:** A menudo hay un dataset con dos círculos concéntricos o dos conjuntos de puntos en forma de anillo.

* **Generación con `scikit-learn` (usando `make_circles`):**




```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt

# Generar un dataset de círculos
X, y = make_circles(n_samples=200, noise=0.1, factor=0.5, random_state=42)

# Visualizar los datos
plt.figure(figsize=(4, 4))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdYlBu, edgecolors='k')
plt.title("Dataset de Círculos")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

print("Forma de los datos (X, y):", X.shape, y.shape)
```


**3. Datasets No Lineales (Espiral):**

Estos datasets son más desafiantes y requieren una red neuronal más compleja para aprender la separación.

* **En TensorFlow Playground:** Puedes ver un dataset con dos espirales entrelazadas.

* **Generación con NumPy:** Necesitaremos crear las coordenadas de los puntos manualmente.



```python
def generate_spiral_data(n_samples=1000, num_classes=2, noise=0.1, turns=4, tightness=1):
    """
    Genera espirales entrelazadas con más vueltas y brazos más juntos.

    Parámetros:
        n_samples: Total de muestras en el dataset.
        num_classes: Número de clases/espirales.
        noise: Nivel de ruido en los datos.
        turns: Número de vueltas que da la espiral.
        tightness: Qué tan juntas están las vueltas (valores bajos las acercan más).
    """
    n = n_samples // num_classes
    X = []
    y = []
    for i in range(num_classes):
        theta = np.linspace(0, turns * np.pi, n)  # Aumentamos el número de vueltas
        r = np.linspace(0, 1, n) ** tightness  # Modifica la separación de los brazos
        x = r * np.cos(theta + i * np.pi) + np.random.randn(n) * noise  # Se añade ruido
        y_ = r * np.sin(theta + i * np.pi) + np.random.randn(n) * noise
        X.append(np.c_[x, y_])
        y.append(np.full(n, i))
    return np.vstack(X), np.hstack(y)

# Generar el dataset de espiral
X, y = generate_spiral_data(n_samples=1000)

# Visualizar los datos
plt.figure(figsize=(4, 4))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdYlBu, edgecolors='k')
plt.title("Dataset de Espiral")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

print("Forma de los datos (X, y):", X.shape, y.shape)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_spiral_data(n_samples=1000, num_classes=2, noise=0.1):
    """Genera espirales entrelazadas con ruido ajustable."""
    n = n_samples // num_classes
    X = []
    y = []
    for i in range(num_classes):
        theta = np.linspace(0, 2 * np.pi, n)  # Ángulo base de la espiral
        r = np.linspace(0, 1, n)  # Radio creciente
        x = r * np.cos(theta + i * np.pi) + np.random.randn(n) * noise  # Se añade ruido
        y_ = r * np.sin(theta + i * np.pi) + np.random.randn(n) * noise
        X.append(np.c_[x, y_])
        y.append(np.full(n, i))
    return np.vstack(X), np.hstack(y)


# Generar el dataset de espiral
X, y = generate_spiral_data(n_samples=1000)

# Visualizar los datos
plt.figure(figsize=(4, 4))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdYlBu, edgecolors='k')
plt.title("Dataset de Espiral")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

print("Forma de los datos (X, y):", X.shape, y.shape)
```

```python
def generate_xor_data(n_samples=1000, noise=0.1):
    """Genera datos estilo XOR."""
    X, y = make_moons(n_samples=n_samples, noise=noise)
    return X, y


# Generar el dataset XOR
X, y = generate_xor_data(n_samples=1000)

# Visualizar los datos
plt.figure(figsize=(4, 4))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdYlBu, edgecolors='k')
plt.title("Dataset XOR")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
def generate_xor_square_data(n_samples=1000, noise=0.1):
    """
    Genera datos estilo XOR en forma de un tablero de ajedrez 2x2.

    Cada cuadrante contiene puntos distribuidos uniformemente con una clase fija,
    y el ruido permite dispersarlos un poco.

    Parámetros:
        n_samples: Total de muestras.
        noise: Cantidad de ruido en la posición de los puntos.
    """
    n = n_samples // 4  # Número de puntos por cuadrante
    X = []
    y = []

    # Centros de cada cuadrante y clase asociada
    quadrants = [
        ((0.5, 0.5), 1),   # Cuadrante superior derecho
        ((-0.5, 0.5), 0),  # Cuadrante superior izquierdo
        ((0.5, -0.5), 0),  # Cuadrante inferior derecho
        ((-0.5, -0.5), 1)  # Cuadrante inferior izquierdo
    ]

    for (cx, cy), label in quadrants:
        # Rellenamos el cuadrante uniformemente en un cuadrado [-0.5, 0.5] y lo desplazamos al centro
        x = np.random.rand(n, 2) - 0.5
        x += np.array([cx, cy])  # Desplazar al cuadrante correspondiente
        X.append(x)
        y.append(np.full(n, label))

    # Añadir ruido a todas las muestras
    X = np.vstack(X) + np.random.randn(n_samples, 2) * noise
    y = np.hstack(y)
    return X, y


# Generar el dataset XOR cuadrado
X, y = generate_xor_square_data(n_samples=1000)

# Visualizar los datos
plt.figure(figsize=(4, 4))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdYlBu, edgecolors='k')
plt.title("Dataset XOR Cuadrado")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

TensorFlow Playground también ofrece datasets con patrones más complejos o combinaciones de los anteriores. Puedes intentar generar estos patrones combinando las ideas anteriores o buscando funciones de generación más específicas en bibliotecas como `scikit-learn`.

**Personalización de los Datasets:**

Puedes ajustar varios parámetros en las funciones de generación para controlar la complejidad, el número de muestras, el nivel de ruido, etc. Esto te permitirá crear datasets específicos para ilustrar diferentes conceptos a tus estudiantes.

* **`n_samples`:** Número de puntos de datos.
* **`n_features`:** Número de características (generalmente 2 para visualización en 2D).
* **`noise`:** Cantidad de ruido en los datos.
* **`factor` (para `make_circles`):** Factor de escala entre los círculos.
* **`random_state`:** Semilla para la generación de números aleatorios, lo que asegura que obtengas los mismos datos cada vez que ejecutes el código.
