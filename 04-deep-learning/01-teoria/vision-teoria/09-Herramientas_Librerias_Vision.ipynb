## 1. OpenCV: procesamiento de imágenes

```python
from pathlib import Path
import cv2
import matplotlib.pyplot as plt

# Leer imagen
image_path = Path("vision/teoria/example.jpg")
if not image_path.exists():
    image_path = Path("example.jpg")

img = cv2.imread(str(image_path))
if img is None:
    raise FileNotFoundError("No se ha podido leer la imagen de ejemplo.")

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Mostrar imagen
plt.imshow(img_rgb)
plt.title("Imagen cargada con OpenCV")
plt.axis('off')
plt.show()
```

## 2. PIL/Pillow: manipulación básica de imágenes

```python
from PIL import Image, ImageOps

# Abrir imagen y convertir a escala de grises
img_pil = Image.open(image_path)
gray_pil = ImageOps.grayscale(img_pil)

plt.figure(figsize=(5,5))
plt.imshow(gray_pil, cmap='gray')
plt.title("Imagen en escala de grises con PIL")
plt.axis('off')
plt.show()
```
