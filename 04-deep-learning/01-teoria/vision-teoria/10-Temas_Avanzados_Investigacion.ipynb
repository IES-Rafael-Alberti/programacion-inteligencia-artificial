## 1. Visión multimodal
- Combina imágenes con texto, audio, sensores u otras fuentes de información.
- Ejemplos:
  - generar descripciones automáticas de imágenes
  - buscar imágenes a partir de texto
  - analizar vídeo junto con audio o subtítulos
- Modelos y herramientas frecuentes:
  - CLIP
  - BLIP
  - OpenCLIP
  - Hugging Face Transformers

### Qué objetivo tendría en clase
Mostrar que la visión ya no trabaja aislada: muchas aplicaciones modernas conectan imagen y lenguaje.

  - SHAP

## 3. Visión en el borde (Edge AI)
- Consiste en ejecutar modelos directamente en dispositivos cercanos a la fuente de datos:
  - móviles
  - cámaras inteligentes
  - Raspberry Pi
  - Jetson
- Ventajas:
  - menor latencia
  - más privacidad
  - menos dependencia de la nube
- Herramientas frecuentes:
  - TensorFlow Lite
  - OpenVINO
  - NVIDIA Jetson
  - Coral TPU

### Qué objetivo tendría en clase
Entender que no basta con que un modelo funcione: también importa dónde y cómo se ejecuta.

## 4. Realidad aumentada y realidad virtual
- Superponen o integran elementos digitales en escenas reales o simuladas.
- Ejemplos:
  - filtros faciales
  - guías visuales en mantenimiento o medicina
  - simuladores inmersivos
- Herramientas frecuentes:
  - OpenCV
  - MediaPipe
  - Unity
  - ARKit / ARCore

### Qué objetivo tendría en clase
Conectar visión por computador con interacción, interfaces y aplicaciones inmersivas.

## 5. Ética, sesgos y responsabilidad
- La visión por computador puede amplificar sesgos del dataset o del contexto de recogida de datos.
- Algunos problemas típicos:
  - reconocimiento facial desigual según grupos de población
  - vigilancia masiva sin consentimiento
  - deepfakes y manipulación visual
  - uso de sistemas con falsa sensación de objetividad

### Qué objetivo tendría en clase
Recordar que un sistema visual no es solo un modelo técnico: también tiene implicaciones sociales, legales y éticas.
