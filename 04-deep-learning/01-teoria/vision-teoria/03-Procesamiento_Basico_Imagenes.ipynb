# Procesamiento Básico de Imágenes
Este notebook cubre operaciones fundamentales de procesamiento de imágenes con OpenCV: umbralización, filtros de suavizado y detección de bordes.

```python
# Librerías necesarias
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Markdown
```

```python
# Carga de imagen (Colab o local)
try:
    from google.colab import files
    uploaded = files.upload()
    img_path = list(uploaded.keys())[0]
except:
    img_path = "/home/jmsa/Imágenes/pose1.jpg"

img = cv2.imread(img_path)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap='gray')
plt.title("Imagen en escala de grises")
plt.axis('off')
plt.show()
```

```python
# Umbralización binaria
_, thresh = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)
plt.imshow(thresh, cmap='gray')
plt.title("Umbralización binaria")
plt.axis('off')
plt.show()
```

```python
# Filtros de suavizado: media y gaussiano
blur_mean = cv2.blur(img_gray, (5,5))
blur_gauss = cv2.GaussianBlur(img_gray, (5,5), 0)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(blur_mean, cmap='gray')
plt.title("Filtro de media")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(blur_gauss, cmap='gray')
plt.title("Filtro gaussiano")
plt.axis('off')
plt.show()
```

```python
# Detección de bordes: Sobel y Canny
sobelx = cv2.Sobel(img_gray, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(img_gray, cv2.CV_64F, 0, 1, ksize=5)
edges_canny = cv2.Canny(img_gray, 100, 200)

plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(sobelx, cmap='gray')
plt.title("Sobel X")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(sobely, cmap='gray')
plt.title("Sobel Y")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(edges_canny, cmap='gray')
plt.title("Canny")
plt.axis('off')
plt.show()
```
