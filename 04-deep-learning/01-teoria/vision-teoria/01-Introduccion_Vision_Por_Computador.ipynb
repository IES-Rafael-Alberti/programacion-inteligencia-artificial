# Introducción a la Visión por Computador
Este notebook introduce los conceptos básicos de visión por computador con ejemplos en Python usando OpenCV y Matplotlib.

```python
# ¿Qué es visión por computador?
print("La visión por computador permite que los ordenadores interpreten y comprendan imágenes del mundo real.")
```

```python
# Carga de imagen (compatible con Colab o Jupyter)
import cv2
import matplotlib.pyplot as plt
from IPython.display import Markdown

try:
    from google.colab import files
    uploaded = files.upload()
    img_path = list(uploaded.keys())[0]
except (ImportError, Exception):
    img_path = 'example-2.jpg'

img_path = "/home/jmsa/Imágenes/pose1.jpg"
img = cv2.imread(img_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # type: ignore
```

```python
# Visualización de imagen
plt.imshow(img_rgb)
plt.title("Ejemplo de imagen cargada")
plt.axis('off')
plt.show()
```

```python
# Análisis básico de imagen
print(f"Dimensiones: {img.shape}")
print(f"Tipo de datos: {img.dtype}")
print(f"Valor medio por canal: {img.mean(axis=(0, 1))}")
```

```python
# Historia de la visión por computador
Markdown('''
### Breve historia de la visión por computador
- **Años 60-70**: Primeros experimentos con reconocimiento de formas.
- **Décadas siguientes**: Avances en segmentación y extracción de características (SIFT, SURF).
- **2012 en adelante**: Deep Learning (AlexNet) supera métodos tradicionales en reconocimiento de imágenes.
''')
```

```python
# Aplicaciones actuales
apps = [
    "✔ Reconocimiento facial",
    "✔ Conducción autónoma",
    "✔ Diagnóstico médico asistido",
    "✔ Control de calidad industrial",
    "✔ Agricultura de precisión",
    "✔ Vigilancia inteligente"
]
print("Aplicaciones destacadas:")
for app in apps:
    print(app)
```

```python
# Comparación visual: original vs desenfocada
img_blur = cv2.GaussianBlur(img_rgb, (25, 25), 0)

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Imagen original")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_blur)
plt.title("Imagen desenfocada")
plt.axis('off')
plt.show()
```
