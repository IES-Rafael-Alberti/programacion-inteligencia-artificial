# Detección facial con dlib y OpenCV
Este notebook muestra cómo detectar rostros y 68 puntos faciales usando la combinación de dlib y OpenCV.

```python
import cv2
import dlib
import matplotlib.pyplot as plt

# Cargar detector y predictor
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")

# Leer imagen
img = cv2.imread("ejemplo.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Detección de rostros
faces = detector(img_rgb)

# Dibujar resultados
for face in faces:
    landmarks = predictor(img_rgb, face)
    for n in range(68):
        x = landmarks.part(n).x
        y = landmarks.part(n).y
        cv2.circle(img_rgb, (x, y), 2, (0, 255, 0), -1)
    cv2.rectangle(img_rgb, (face.left(), face.top()), (face.right(), face.bottom()), (255, 0, 0), 2)

# Mostrar imagen
plt.figure(figsize=(8,6))
plt.imshow(img_rgb)
plt.axis('off')
plt.title("Rostro con puntos faciales detectados")
plt.show()
```
