# Visión 3D con OpenCV
Este notebook introduce los conceptos básicos de visión estéreo, mapas de disparidad y reconstrucción 3D usando OpenCV. Requiere dos imágenes estéreo (izquierda y derecha).

```python
# Instalar OpenCV si es necesario
#!pip install opencv-python opencv-contrib-python
```

```python
# Selección de imágenes estéreo
from pathlib import Path

left_path = Path("stereo_left.jpg")
right_path = Path("stereo_right.jpg")

try:
    from google.colab import files
    uploaded = files.upload()
except Exception:
    uploaded = {}

print(f"Izquierda: {left_path} | existe={left_path.exists()}")
print(f"Derecha: {right_path} | existe={right_path.exists()}")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Leer imágenes
imgL = cv2.imread(str(left_path), cv2.IMREAD_GRAYSCALE)
imgR = cv2.imread(str(right_path), cv2.IMREAD_GRAYSCALE)

if imgL is None or imgR is None:
    raise FileNotFoundError("No se han podido leer las imágenes estéreo. Usa los nombres stereo_left.jpg y stereo_right.jpg o ajusta las rutas.")

# Mostrar imágenes
plt.figure(figsize=(10,4))
plt.subplot(1, 2, 1)
plt.imshow(imgL, cmap='gray')
plt.title('Imagen Izquierda')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(imgR, cmap='gray')
plt.title('Imagen Derecha')
plt.axis('off')
plt.show()
```

```python
# Calcular mapa de disparidad
stereo = cv2.StereoBM_create(numDisparities=64, blockSize=15)
disparity = stereo.compute(imgL, imgR)

# Mostrar el mapa de disparidad
plt.imshow(disparity, 'gray')
plt.title('Mapa de disparidad')
plt.axis('off')
plt.show()
```
