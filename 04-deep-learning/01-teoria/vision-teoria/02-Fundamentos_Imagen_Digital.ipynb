# Fundamentos de Imagen Digital
Este notebook presenta conceptos básicos sobre imágenes digitales y cómo manipularlas usando Python y OpenCV.

```python
# Librerías necesarias
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Markdown
```

```python
# Carga de imagen (Colab o local)
try:
    from google.colab import files
    uploaded = files.upload()
    img_path = list(uploaded.keys())[0]
except (ImportError, Exception):
    img_path = "example.jpg"

img = cv2.imread("/home/jmsa/Imágenes/pose1.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title("Imagen original (RGB)")
plt.axis('off')
plt.show()
```

```python
# Conversión a escala de grises y HSV
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

plt.figure(figsize=(12,4))
plt.subplot(1, 2, 1)
plt.imshow(img_gray, cmap='gray')
plt.title("Escala de grises")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB))
plt.title("HSV")
plt.axis('off')
plt.show()
```

```python
# Recorte y rotación
h, w = img_rgb.shape[:2]
crop = img_rgb[h//4:3*h//4, w//4:3*w//4]

M = cv2.getRotationMatrix2D((w//2, h//2), 45, 1.0)
rotated = cv2.warpAffine(img_rgb, M, (w, h))

plt.figure(figsize=(12,4))
plt.subplot(1, 2, 1)
plt.imshow(crop)
plt.title("Imagen recortada")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(rotated)
plt.title("Imagen rotada 45°")
plt.axis('off')
plt.show()
```

```python
# Análisis de histograma de color
color = ('b','g','r')
plt.figure()
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr, color = col)
    plt.xlim([0,256])
plt.title("Histograma por canal (BGR)")
plt.show()
```
