# Filtros Clásicos en Visión por Computador
Este notebook explora filtros comunes y relevantes para preprocesamiento y mejora de imágenes, como Sobel, Laplaciano, GaussianBlur, Bilateral y Median Blur.

```python
# Librerías necesarias
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Carga de imagen en escala de grises
try:
    from google.colab import files
    uploaded = files.upload()
    img_path = list(uploaded.keys())[0]
except:
    img_path = '/home/jmsa/Imágenes/pose1.jpg'

img = cv2.imread(img_path)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap='gray')
plt.title("Imagen original en escala de grises")
plt.axis('off')
plt.show()
```

```python
# Filtro Sobel (X y Y)
sobelx = cv2.Sobel(img_gray, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(img_gray, cv2.CV_64F, 0, 1, ksize=3)

plt.figure(figsize=(10,4))
plt.subplot(1, 2, 1)
plt.imshow(sobelx, cmap='gray')
plt.title("Sobel X")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sobely, cmap='gray')
plt.title("Sobel Y")
plt.axis('off')
plt.show()
```

```python
# Filtro Laplaciano
laplacian = cv2.Laplacian(img_gray, cv2.CV_64F)
plt.imshow(laplacian, cmap='gray')
plt.title("Laplaciano")
plt.axis('off')
plt.show()
```

```python
# Filtro Gaussiano
gaussian = cv2.GaussianBlur(img_gray, (7, 7), 0)
plt.imshow(gaussian, cmap='gray')
plt.title("Gaussian Blur")
plt.axis('off')
plt.show()
```

```python
# Filtro de Mediana
median = cv2.medianBlur(img_gray, 5)
plt.imshow(median, cmap='gray')
plt.title("Median Blur")
plt.axis('off')
plt.show()
```

```python
# Filtro Bilateral
bilateral = cv2.bilateralFilter(img_gray, 9, 75, 75)
plt.imshow(bilateral, cmap='gray')
plt.title("Bilateral Filter")
plt.axis('off')
plt.show()
```
