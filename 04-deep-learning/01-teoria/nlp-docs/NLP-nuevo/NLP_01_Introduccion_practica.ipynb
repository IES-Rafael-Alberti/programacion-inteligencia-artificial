# NLP Fase 1 — Introducción práctica

En este notebook vamos a ver, paso a paso, cómo pasar de texto a números y cómo entrenar un primer clasificador sencillo.


## 1. Dataset de ejemplo

Empezamos con un conjunto pequeño de frases para centrarnos en el pipeline y no en el tamaño del dataset.


```python
import pandas as pd

texts = [
    "Me encanta este curso",
    "Odio este producto",
    "Este curso es muy bueno",
    "No me gusta nada",
    "Me gusta bastante"
]

labels = [1, 0, 1, 0, 1]

df = pd.DataFrame({"text": texts, "label": labels})
df
```

## 2. Preprocesado

Convertimos a minúsculas y eliminamos signos básicos para simplificar el ejemplo.


```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^a-záéíóúüñ ]", "", text)
    return text

df["clean_text"] = df["text"].apply(clean_text)
df
```

## 3. Bag of Words

Ahora convertimos el texto en una representación numérica basada en conteos.


```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X_bow = vectorizer.fit_transform(df["clean_text"])

print("Vocabulario:")
print(vectorizer.get_feature_names_out())
print("\nMatriz Bag of Words:")
print(X_bow.toarray())
```

## 4. TF-IDF

Bag of Words cuenta palabras. TF-IDF intenta además ponderar su importancia.


```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(df["clean_text"])

print("Vocabulario:")
print(tfidf.get_feature_names_out())
print("\nMatriz TF-IDF:")
print(X_tfidf.toarray())
```

## 5. Primer modelo

Entrenamos una regresión logística sobre la representación TF-IDF.


```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_tfidf, df["label"])

preds = model.predict(X_tfidf)
print("Predicciones:", preds)
```

## 6. Evaluación básica

Empezamos con accuracy y matriz de confusión.


```python
from sklearn.metrics import accuracy_score, confusion_matrix

print("Accuracy:", accuracy_score(df["label"], preds))
print("Matriz de confusión:")
print(confusion_matrix(df["label"], preds))
```

## 7. Interpretación rápida del modelo

Podemos inspeccionar qué palabras parecen empujar hacia una clase u otra.


```python
feature_names = tfidf.get_feature_names_out()
coefficients = model.coef_[0]

importance = pd.DataFrame({
    "palabra": feature_names,
    "peso": coefficients
}).sort_values("peso", ascending=False)

importance
```
