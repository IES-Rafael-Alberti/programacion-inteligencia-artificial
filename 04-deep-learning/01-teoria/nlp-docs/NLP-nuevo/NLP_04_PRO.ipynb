# NLP 04 PRO — Comparación real de modelos

## 1. Dataset

```
import pandas as pd
texts = [
 'me encanta este curso',
 'odio este producto',
 'me gusta bastante',
 'no me gusta nada',
 'este curso es increíble',
 'producto terrible'
]
labels = [1,0,1,0,1,0]
df = pd.DataFrame({'text':texts,'label':labels})
df
```

## 2. TF-IDF

```
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

tfidf = TfidfVectorizer()
X = tfidf.fit_transform(df['text'])
model_tfidf = LogisticRegression()
model_tfidf.fit(X, df['label'])
preds_tfidf = model_tfidf.predict(X)
acc_tfidf = accuracy_score(df['label'], preds_tfidf)
acc_tfidf
```

## 3. LSTM

```
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer()
tokenizer.fit_on_texts(df['text'])
seq = tokenizer.texts_to_sequences(df['text'])
padded = pad_sequences(seq)

model_lstm = tf.keras.Sequential([
    tf.keras.layers.Embedding(1000, 16),
    tf.keras.layers.LSTM(32),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model_lstm.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model_lstm.fit(padded, df['label'], epochs=5, verbose=0)

preds_lstm = (model_lstm.predict(padded) > 0.5).astype(int)
acc_lstm = accuracy_score(df['label'], preds_lstm)
acc_lstm
```

## 4. BERT

```
from transformers import pipeline
clf = pipeline('sentiment-analysis')
results = clf(df['text'].tolist())
results
```

## 5. Comparación

```
print('TF-IDF accuracy:', acc_tfidf)
print('LSTM accuracy:', acc_lstm)
# TODO: adaptar salida de BERT a etiquetas y calcular accuracy
```
