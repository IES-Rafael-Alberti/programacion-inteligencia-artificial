# Notebook 2: Mitigación del Overfitting

Objetivo: aplicar técnicas reales.

```
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

x_train = x_train / 255.0
x_test = x_test / 255.0

x_train = x_train[..., None]
x_test = x_test[..., None]
```

```
def build_model(dropout=False):
    model = models.Sequential()

    model.add(layers.Conv2D(64, (3,3), activation='relu', input_shape=(28,28,1)))
    model.add(layers.Conv2D(64, (3,3), activation='relu'))
    model.add(layers.MaxPooling2D())

    if dropout:
        model.add(layers.Dropout(0.5))

    model.add(layers.Flatten())
    model.add(layers.Dense(128, activation='relu'))

    if dropout:
        model.add(layers.Dropout(0.5))

    model.add(layers.Dense(10, activation='softmax'))

    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    return model
```

```
model_no_reg = build_model(dropout=False)
history_no_reg = model_no_reg.fit(x_train, y_train,
                                 epochs=10,
                                 validation_split=0.2,
                                 verbose=0)
```

```
model_reg = build_model(dropout=True)
history_reg = model_reg.fit(x_train, y_train,
                           epochs=10,
                           validation_split=0.2,
                           verbose=0)
```

```
plt.figure()
plt.plot(history_no_reg.history['val_loss'], label='sin regularización')
plt.plot(history_reg.history['val_loss'], label='con dropout')
plt.legend()
plt.title("Comparación Overfitting")
plt.show()
```
