## Mini-guia Keras (resumen docente)

### Que es Keras
Keras es una API de alto nivel para construir y entrenar redes neuronales de forma clara.

Flujo tipico:
1. Preparar datos
2. Definir red
3. Compilar (`loss`, `optimizer`, `metrics`)
4. Entrenar (`fit`)
5. Evaluar (`evaluate`)

### Formas de construir modelos
- `Sequential`: pila lineal de capas.
- API funcional: mayor flexibilidad (multi-entrada/salida).
- Subclassing: control avanzado.

### Capas frecuentes
- `Dense` (MLP)
- `Dropout` (regularizacion)
- `BatchNormalization` (estabilidad)
- `Conv2D/MaxPooling2D` (imagenes)

### Activaciones
- Ocultas: `relu` (normalmente)
- Salida binaria: `sigmoid`
- Salida multiclase: `softmax`
- Regresion: `linear`

### Perdida y metrica (regla rapida)
- Binaria: `binary_crossentropy` + `accuracy`
- Multiclase: `sparse_categorical_crossentropy` o `categorical_crossentropy`
- Regresion: `mse`/`mae`

### Regularizacion y control
- `Dropout`
- L2 (`kernel_regularizer`)
- `EarlyStopping`
- `ReduceLROnPlateau`

### Batches y epochs
- `batch_size`: muestras por actualizacion
- `epoch`: pasada completa por dataset

Regla inicial util: `batch_size=32` o `64`.

```

import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers
```


## 1. Carga del dataset Fashion‑MNIST


```

(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

print("Train:", x_train.shape, y_train.shape)
print("Test :", x_test.shape, y_test.shape)
```


## 2. Exploración visual inicial


```

class_names = [
    "T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
    "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"
]

plt.figure(figsize=(8, 8))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.imshow(x_train[i], cmap="gray")
    plt.title(class_names[y_train[i]])
    plt.axis("off")
plt.show()
```


## 3. Construcción del problema binario


```

shoe_classes = [5, 7, 9]

y_train_bin = np.isin(y_train, shoe_classes).astype(int)
y_test_bin = np.isin(y_test, shoe_classes).astype(int)
```


## 4. Preprocesado


```

x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

x_train = x_train.reshape(-1, 28 * 28)
x_test = x_test.reshape(-1, 28 * 28)
```


## 5. Definición del modelo


```

model = keras.Sequential([
    layers.Dense(128, activation="relu", input_shape=(784,)),
    layers.Dense(1, activation="sigmoid")
])

model.summary()
```


## 6. Compilación


```

model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
```


## 7. Entrenamiento


```

history = model.fit(
    x_train,
    y_train_bin,
    epochs=10,
    batch_size=128,
    validation_split=0.2
)
```


## 8. Evaluación


```

model.evaluate(x_test, y_test_bin)
```


## 9. Curvas de entrenamiento


```

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history["loss"], label="Train")
plt.plot(history.history["val_loss"], label="Validation")
plt.title("Loss")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history["accuracy"], label="Train")
plt.plot(history.history["val_accuracy"], label="Validation")
plt.title("Accuracy")
plt.legend()

plt.show()
```
