# 🧪 Visualización: Atención Encoder-Decoder (Cross-Attention)

Este notebook te permite observar cómo varía la atención encoder-decoder en un modelo Transformer en diferentes fases de generación de texto.

Veremos cómo cambia el foco de atención del decodificador hacia distintas partes de la entrada a medida que se generan las palabras de salida.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Palabras de entrada y salida simuladas
entrada = ["El", "gato", "duerme"]
salida = ["The", "cat", "sleeps"]

# Simular pesos de atención encoder-decoder en cada paso de la salida
# Cada fila representa la atención del token de salida hacia la entrada
attention_weights = np.array([
    [0.6, 0.3, 0.1],  # The -> mostly 'El'
    [0.2, 0.7, 0.1],  # cat -> mostly 'gato'
    [0.1, 0.2, 0.7]   # sleeps -> mostly 'duerme'
])
```

## 🎯 Visualizar atención por palabra generada

```python
# Mostrar cada paso como mapa de calor
for i, palabra_salida in enumerate(salida):
    plt.figure(figsize=(6, 1.5))
    sns.heatmap([attention_weights[i]], annot=True, cmap="YlGnBu",
                xticklabels=entrada, yticklabels=[palabra_salida], cbar=True)
    plt.title(f"Atención del decoder al generar: '{palabra_salida}'")
    plt.xlabel("Entrada (encoder)")
    plt.ylabel("Salida (decoder)")
    plt.tight_layout()
    plt.show()
```
