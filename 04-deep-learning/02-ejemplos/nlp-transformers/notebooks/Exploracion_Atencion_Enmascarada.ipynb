# 🧪 Exploración: Atención Enmascarada (Masked Self-Attention)

En esta actividad podrás visualizar cómo se aplica una máscara en el mecanismo de self-attention del decoder de un Transformer.

La máscara evita que una palabra vea las siguientes posiciones en la secuencia durante la generación paso a paso.

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_mask(seq_len):
    return np.triu(np.ones((seq_len, seq_len)), k=1) * -1e9
```

## 🔧 Parámetro configurable
- Cambia la longitud de la secuencia (por ejemplo, 4, 6, 10) y vuelve a ejecutar la celda siguiente para ver la máscara generada.

```python
# Cambia este valor:
seq_len = 6

# Generar máscara triangular superior (futura bloqueada)
mask = generate_mask(seq_len)

# Visualizar la máscara
plt.figure(figsize=(6, 5))
plt.imshow(mask, cmap="gray", aspect='equal')
plt.title(f"Máscara de atención (seq_len = {seq_len})")
plt.xlabel("Key")
plt.ylabel("Query")
plt.colorbar(label="Score")
plt.show()

# Mostrar matriz como números para inspección
import pandas as pd
df = pd.DataFrame(mask, columns=[f"K{i+1}" for i in range(seq_len)], index=[f"Q{i+1}" for i in range(seq_len)])
df
```
