# 🧪 Exploración: Codificación Posicional (sinusoidal)

En esta actividad podrás experimentar con la codificación posicional que usan los modelos Transformer.

Puedes cambiar los parámetros para observar cómo cambian los vectores que se suman a los embeddings de entrada.

```python
import numpy as np
import matplotlib.pyplot as plt

def positional_encoding(pos, d_model):
    PE = np.zeros((pos, d_model))
    for p in range(pos):
        for i in range(0, d_model, 2):
            angle = p / (10000 ** ((2 * i)/d_model))
            PE[p, i] = np.sin(angle)
            if i + 1 < d_model:
                PE[p, i+1] = np.cos(angle)
    return PE
```

## 🔧 Parámetros configurables
- Cambia estos valores y vuelve a ejecutar la siguiente celda para ver el efecto.
- Prueba distintas combinaciones de posiciones y dimensiones del embedding.

```python
# Puedes modificar estos valores:
num_positions = 50      # Número de posiciones (tokens en la secuencia)
embedding_dim = 16      # Dimensión del embedding (ej. 8, 16, 32, 64...)

# Calcular codificación posicional
pe_matrix = positional_encoding(num_positions, embedding_dim)

# Visualizar
plt.figure(figsize=(12, 6))
plt.imshow(pe_matrix.T, cmap="viridis", aspect="auto")
plt.title(f"Codificación Posicional - {num_positions} posiciones, {embedding_dim} dimensiones")
plt.xlabel("Posición en la secuencia")
plt.ylabel("Dimensión del embedding")
plt.colorbar(label="Valor")
plt.show()
```
