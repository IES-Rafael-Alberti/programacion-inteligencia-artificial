# 🧠 Práctica: Mecanismo de Atención con Dos Cabezas

En esta práctica vamos a simular dos cabezas de atención con diferentes matrices de pesos. Veremos cómo cada cabeza produce una representación distinta y cómo se combinan al final.

```python
# Paso 1: Definimos los embeddings
import numpy as np

words = ["El", "gato", "duerme"]
embeddings = {
    "El": np.array([1, 0]),
    "gato": np.array([0, 1]),
    "duerme": np.array([1, 1])
}
```

## 🔹 Cabeza 1

```python
# Matrices de pesos para cabeza 1
Wq1 = np.array([[1, 0], [0, 1]])
Wk1 = np.array([[0, 1], [1, 0]])
Wv1 = np.array([[1, 1], [0, 1]])

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()

def compute_attention(words, Wq, Wk, Wv):
    Q, K, V = {}, {}, {}
    for word, vec in embeddings.items():
        Q[word] = vec @ Wq
        K[word] = vec @ Wk
        V[word] = vec @ Wv
    
    scores = np.zeros((len(words), len(words)))
    for i, qw in enumerate(words):
        for j, kw in enumerate(words):
            scores[i, j] = np.dot(Q[qw], K[kw])
    
    weights = np.apply_along_axis(softmax, 1, scores)

    outputs = []
    for i in range(len(words)):
        weighted_sum = sum(weights[i, j] * V[words[j]] for j in range(len(words)))
        outputs.append(weighted_sum)
    return outputs, weights

outputs1, weights1 = compute_attention(words, Wq1, Wk1, Wv1)
outputs1
```

## 🔹 Cabeza 2

```python
# Matrices de pesos para cabeza 2
Wq2 = np.array([[0, 1], [1, 0]])
Wk2 = np.array([[1, 1], [1, 0]])
Wv2 = np.array([[0, 1], [1, 1]])

outputs2, weights2 = compute_attention(words, Wq2, Wk2, Wv2)
outputs2
```

## 🔗 Concatenación de ambas cabezas

```python
# Concatenar las salidas de ambas cabezas
final_outputs = [np.concatenate((o1, o2)) for o1, o2 in zip(outputs1, outputs2)]
final_outputs
```
