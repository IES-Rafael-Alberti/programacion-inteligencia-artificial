# 80 - Métricas de Segmentación — Base

Cálculo de **IoU** y **Dice Score** sobre predicciones sintéticas para evaluar segmentación.

```

import numpy as np

def iou(y_true, y_pred):
    inter = np.logical_and(y_true,y_pred).sum()
    union = np.logical_or(y_true,y_pred).sum()
    return inter/union if union>0 else 1.0

def dice(y_true,y_pred):
    inter = np.logical_and(y_true,y_pred).sum()
    return (2*inter)/(y_true.sum()+y_pred.sum()+1e-8)

# Ejemplo sintético
rng=np.random.default_rng(42)
mask_true=(rng.random((64,64))>0.7).astype(int)
mask_pred=(rng.random((64,64))>0.75).astype(int)

print("IoU:", round(iou(mask_true,mask_pred),3))
print("Dice:", round(dice(mask_true,mask_pred),3))
```
