```
try:
    import torch, torchvision; from torchvision import models
    HAS_TORCH=True; print('✅ PyTorch disponible')
except Exception as e:
    HAS_TORCH=False; print('⚠️ PyTorch no disponible:', e)
if HAS_TORCH:
    device=torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    try:
        model=models.resnet18(weights=models.ResNet18_Weights.DEFAULT).to(device).eval()
        print('Modelo ResNet18 cargado')
    except Exception as e:
        model=None; print('⚠️ No se pudo cargar modelo:', e)
    if model:
        target_layer=model.layer4[1].conv2
        activations,gradients={},{}
        def f_hook(m,i,o): activations['value']=o.detach()
        def b_hook(m,gi,go): gradients['value']=go[0].detach()
        target_layer.register_forward_hook(f_hook)
        target_layer.register_backward_hook(b_hook)
        print('Hooks registrados')
```
