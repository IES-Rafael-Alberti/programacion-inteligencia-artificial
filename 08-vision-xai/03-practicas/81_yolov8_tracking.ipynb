## (Opcional) Generación de vídeo sintético con OpenCV

```
try:
    import cv2, numpy as np
    HAS_CV=True
except Exception as e:
    HAS_CV=False; print('⚠️ OpenCV no disponible:', e)
if HAS_CV:
    h,w=240,320; path='/mnt/data/synth_video.mp4'
    fourcc=cv2.VideoWriter_fourcc(*'mp4v'); vw=cv2.VideoWriter(path,fourcc,15,(w,h))
    for t in range(60):
        fr=np.zeros((h,w,3),np.uint8)
        cv2.rectangle(fr,(20+t,40),(60+t,80),(0,255,0),-1)
        cv2.circle(fr,(220-t,150),15,(255,255,0),-1)
        vw.write(fr)
    vw.release(); print('🎥 Guardado', path)
else:
    print('Instala opencv-python para generar un vídeo sintético.')
```
