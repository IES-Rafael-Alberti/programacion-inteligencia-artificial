# 84 - LIME en NLP (Texto) — Base

```python
try:
    import lime
    from lime.lime_text import LimeTextExplainer
    HAS_LIME=True; print('✅ LIME disponible')
except Exception as e:
    HAS_LIME=False; print('⚠️ LIME no disponible:', e)
from sklearn.pipeline import make_pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
train_texts=['me gusta este producto','odio esta película','excelente calidad','muy malo','recomendado']
train_labels=[1,0,1,0,1]
model=make_pipeline(TfidfVectorizer(),LogisticRegression()).fit(train_texts,train_labels)
print('Modelo entrenado')
if HAS_LIME:
    explainer=LimeTextExplainer(class_names=['neg','pos'])
    exp=explainer.explain_instance('me encanta este producto', model.predict_proba, num_features=5)
    print('Explicación ejemplo generada')
```
