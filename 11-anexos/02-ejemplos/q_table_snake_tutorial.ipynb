## 🧠 ¿Por qué funcionan los refuerzos y castigos?

### ✅ Recompensas (Refuerzos positivos)
- Indican que una acción fue útil.
- Refuerzan la probabilidad de repetir esa acción.
- En Snake: comer comida → +1 recompensa → el agente aprende a buscarla.
- Ejemplo: si ir a la derecha lleva a comida, se aumenta Q(s, derecha).

### ❌ Castigos (Refuerzos negativos)
- Indican que una acción fue perjudicial.
- Disminuyen la probabilidad de repetirla.
- En Snake: chocar contra un muro o su cuerpo → -1 castigo.
- Así el agente aprende a evitar esas situaciones peligrosas.

### 🔁 Retroalimentación continua
- Aprender solo al final (por morir o ganar) es lento y confuso.
- Mejor: señales frecuentes e intermedias.
- En Snake:
    - 🟩 +1 por comida → motivación positiva
    - 🟥 -1 por morir → corrección clara
    - ➖  0 por sobrevivir sin novedad → neutral

### 📌 Conclusión
- Se basa en ensayo y error, como el aprendizaje humano.
- Usa recompensas y castigos para modificar el comportamiento.
- El agente aprende progresivamente una estrategia óptima.
- Funcionan porque:
  - Proveen una señal inmediata sobre las decisiones.
  - Incentivan comportamientos deseados.
  - Penalizan errores de forma directa y útil.
