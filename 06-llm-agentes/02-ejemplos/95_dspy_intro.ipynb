## Ampliación progresiva

DSPy no consiste solo en escribir prompts, sino en declarar una tarea, probar ejemplos y mejorar la estrategia. Aquí lo simulamos sin depender de una API externa.

Esta celda define ejemplos de evaluación y estilos de prompt. Después calcula una puntuación para comparar qué formulación produce respuestas más alineadas con los términos esperados.


```
examples = [
    {"question": "qué es overfitting", "expected_terms": ["generaliza", "entrenamiento"]},
    {"question": "define gradiente descendente", "expected_terms": ["optimizar", "pérdida"]},
]

def build_prompt(question, style="breve"):
    templates = {
        "breve": "Responde en una frase: {question}",
        "didactico": "Explica para alumnado de FP, con ejemplo: {question}",
        "evaluacion": "Da definición, ejemplo y error común: {question}",
    }
    return templates[style].format(question=question.strip())

for style in ["breve", "didactico", "evaluacion"]:
    print(style, "->", build_prompt("qué es overfitting", style))
```

Esta celda simula respuestas de un modelo y evalúa distintos estilos de prompt. El objetivo es ver cómo DSPy permite comparar alternativas con una métrica, aunque aquí se haga de forma determinista.


```
def simulated_answer(prompt):
    text = prompt.lower()
    if "overfitting" in text:
        return "El overfitting ocurre cuando el modelo memoriza el entrenamiento y no generaliza bien."
    if "gradiente" in text:
        return "El gradiente descendente permite optimizar parámetros reduciendo una función de pérdida."
    return "Respuesta genérica."

def score_answer(answer, expected_terms):
    answer = answer.lower()
    return sum(term in answer for term in expected_terms) / len(expected_terms)

def evaluate_style(style):
    scores = []
    for ex in examples:
        prompt = build_prompt(ex["question"], style)
        answer = simulated_answer(prompt)
        scores.append(score_answer(answer, ex["expected_terms"]))
    return sum(scores) / len(scores)

for style in ["breve", "didactico", "evaluacion"]:
    print(style, evaluate_style(style))
```
