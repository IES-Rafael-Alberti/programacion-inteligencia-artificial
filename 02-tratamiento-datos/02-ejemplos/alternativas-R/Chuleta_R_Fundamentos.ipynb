## 1️⃣ Asignación y tipos de datos

```r
# Asignaciones en R
x <- 10      # asignación con flecha izquierda
20 -> y      # asignación con flecha derecha
z = 30       # asignación con signo igual
x; y; z
```

### Tipos de datos básicos en R

| Tipo | Ejemplo | Comprobación |
|------|----------|--------------|
| Numérico | `x <- 3.14` | `is.numeric(x)` |
| Entero | `x <- 10L` | `is.integer(x)` |
| Lógico | `x <- TRUE` | `is.logical(x)` |
| Cadena | `x <- "Hola"` | `is.character(x)` |
| Complejo | `x <- 3 + 2i` | `is.complex(x)` |

```r
num <- 3.14
ent <- 10L
logi <- FALSE
cad <- "IA"
comp <- 2 + 3i

list(numerico=is.numeric(num), entero=is.integer(ent), logico=is.logical(logi), caracter=is.character(cad), complejo=is.complex(comp))
```

## 2️⃣ Operadores aritméticos, lógicos y de comparación

```r
# Aritméticos
a <- 10; b <- 3
c(suma = a + b,
  resta = a - b,
  multiplicacion = a * b,
  division = a / b,
  potencia = a ^ b,
  modulo = a %% b,
  division_entera = a %/% b)

# Comparación y lógicos
x <- 5; y <- 8
c(igual = x == y,
  distinto = x != y,
  mayor = x > y,
  menor = x < y,
  y_logico = (x > 0) & (y > 0),
  o_logico = (x > 0) | (y < 0),
  negacion = !TRUE)
```

## 3️⃣ Vectores y matrices

```r
# Vector numérico
v <- c(1, 2, 3, 4, 5)
v[1]; length(v); sum(v); mean(v)

# Matriz 3x3
m <- matrix(1:9, nrow=3, ncol=3)
m
t(m)     # transpuesta
```

## 4️⃣ Data Frames

```r
df <- data.frame(
  nombre = c("Ana", "Luis", "Marta"),
  edad = c(20, 22, 19),
  nota = c(8.5, 7.3, 9.1)
)
df
df$edad
mean(df$nota)
```

## 5️⃣ Lectura y guardado de ficheros CSV y Parquet

```r
library(readr)
library(arrow)

# Crear ejemplo y guardar CSV
write_csv(df, "datos.csv")
read_csv("datos.csv")

# Guardar y leer formato Parquet
write_parquet(df, "datos.parquet")
read_parquet("datos.parquet")
```

## 6️⃣ Funciones predefinidas y de usuario

```r
# Funciones predefinidas
x <- c(2, 4, 6, 8)
c(suma=sum(x), media=mean(x), mediana=median(x), desviacion=sd(x))

# Función definida por el usuario
cuadrado <- function(n) {
  return(n^2)
}
cuadrado(5)
```

## 8️⃣ Estadística básica

```r
set.seed(123)
datos <- rnorm(100, mean=50, sd=10)

summary(datos)
var(datos)
sd(datos)
hist(datos, col="lightblue", main="Histograma de datos", xlab="Valor")
```

## 9️⃣ Gráficos básicos

```r
# Scatterplot
x <- 1:10; y <- x^2
plot(x, y, type="b", col="blue", pch=19, main="Relación cuadrática", xlab="x", ylab="y^2")

# Boxplot
boxplot(datos, main="Boxplot de datos aleatorios", col="lightgreen")

# Gráfico con ggplot2
dfg <- data.frame(x=x, y=y)
ggplot(dfg, aes(x, y)) + geom_line(color="blue") + geom_point(color="red", size=3) +
  labs(title="Gráfico con ggplot2", x="x", y="y")
```
