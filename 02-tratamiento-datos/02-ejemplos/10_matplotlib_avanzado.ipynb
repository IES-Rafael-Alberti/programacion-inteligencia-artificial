## 1. Dataset sintético

```python
import numpy as np, pandas as pd
import matplotlib.pyplot as plt

rng = np.random.default_rng(0)
n = 300
df = pd.DataFrame({
    "edad": rng.integers(18, 70, size=n),
    "ingresos": rng.normal(26000, 9000, size=n).round(2),
    "ciudad": rng.choice(["Cádiz","Sevilla","Málaga","Córdoba"], size=n),
    "satisfaccion": rng.integers(1, 6, size=n),
})
df.head()
```

## 2. Histograma

```python
plt.figure()
plt.hist(df["ingresos"].dropna(), bins=30)
plt.title("Distribución de ingresos")
plt.xlabel("Ingresos"); plt.ylabel("Frecuencia")
plt.show()
```

## 3. Boxplot por categoría

```python
plt.figure()
df.boxplot(column="ingresos", by="ciudad", grid=False)
plt.title("Ingresos por ciudad")
plt.suptitle("")
plt.xlabel("Ciudad"); plt.ylabel("Ingresos")
plt.show()
```

## 4. Dispersión con recta de ajuste (polyfit)

```python
x = df["edad"].values
y = df["ingresos"].values

# Ajuste lineal y = a*x + b
a, b = np.polyfit(x, y, 1)

plt.figure()
plt.scatter(x, y, alpha=0.6)
plt.plot(np.sort(x), a*np.sort(x) + b, linewidth=2)
plt.title("Edad vs Ingresos con ajuste lineal")
plt.xlabel("Edad"); plt.ylabel("Ingresos")
plt.show()
```

## 5. Anotaciones y guardado

```python
media_ing = df["ingresos"].mean()

plt.figure()
plt.hist(df["ingresos"], bins=25)
plt.axvline(media_ing, linestyle="--")
plt.text(media_ing, plt.ylim()[1]*0.9, f"media = {media_ing:.0f}")
plt.title("Histograma con línea de media")
plt.xlabel("Ingresos"); plt.ylabel("Frecuencia")
plt.tight_layout()
out_png = "/mnt/data/fig_hist_ingresos.png"
plt.savefig(out_png, dpi=120)
out_png
```
