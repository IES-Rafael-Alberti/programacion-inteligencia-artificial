
# 🧠 Chuleta práctica de **Altair**
Altair es una librería declarativa de Python para crear gráficos estadísticos y realizar **EDA (Análisis Exploratorio de Datos)**.

📘 Documentación oficial: [https://altair-viz.github.io/](https://altair-viz.github.io/)

En esta chuleta aprenderás las funciones y componentes esenciales de Altair, con ejemplos ejecutables directamente.


## ⚙️ Instalación y configuración

```python

%pip install -q altair vega_datasets pandas

import altair as alt
from vega_datasets import data
import pandas as pd

alt.data_transformers.disable_max_rows()
print("Altair versión:", alt.__version__)
```

## 📊 Dataset de ejemplo

```python

df = data.cars()  # dataset integrado
df.head()
```

## 🪄 alt.Chart() — Crear un gráfico base


`alt.Chart(data)` crea el objeto gráfico base sobre el que añadimos elementos con `mark_*()` y `encode()`.

**Parámetros principales:**
- `data`: DataFrame o conjunto de datos.
- `title`: Título opcional.

📘 Ejemplo de gráfico simple con Altair.


```python

chart = alt.Chart(df, title="Ejemplo básico Altair")
chart.mark_point().encode(
    x="Horsepower",
    y="Miles_per_Gallon"
)
```

## 🎨 mark_*() — Tipos de marcas


Altair usa **marcas** para definir el tipo de gráfico.

| Método | Descripción |
|:--------|:-------------|
| `mark_point()` | Diagrama de dispersión |
| `mark_bar()` | Gráfico de barras |
| `mark_line()` | Línea |
| `mark_area()` | Área |
| `mark_circle()` | Círculos (variante de punto) |

📘 Ejemplo: comparación de `mark_bar` y `mark_point`.


```python

alt.Chart(df).mark_bar().encode(
    x="Origin",
    y="mean(Horsepower)"
)
```

```python

alt.Chart(df).mark_point(color="red").encode(
    x="Horsepower",
    y="Miles_per_Gallon",
    tooltip=["Name", "Horsepower", "Miles_per_Gallon"]
)
```

## 🎯 encode() — Asignar variables a canales visuales


`encode()` asocia las variables del dataset con canales visuales como ejes, color, tamaño o forma.

| Canal | Uso |
|:------|:----|
| `x`, `y` | Ejes principales |
| `color` | Color por categoría |
| `size` | Tamaño de puntos |
| `shape` | Forma |
| `tooltip` | Información al pasar el cursor |

📘 Ejemplo: gráfico de dispersión con color y tamaño.


```python

alt.Chart(df).mark_circle(opacity=0.6).encode(
    x="Horsepower",
    y="Miles_per_Gallon",
    color="Origin",
    size="Acceleration",
    tooltip=["Name","Horsepower","Miles_per_Gallon","Origin"]
)
```

## 🔁 transform_*() — Agregaciones y filtros


Altair permite aplicar transformaciones dentro del gráfico sin modificar el DataFrame.

| Transformación | Descripción |
|:----------------|:-------------|
| `transform_filter(expr)` | Filtra datos |
| `transform_aggregate()` | Agrega valores |
| `transform_bin()` | Agrupa en intervalos |
| `transform_calculate()` | Crea columnas derivadas |

📘 Ejemplo: top 10 orígenes por caballos promedio.


```python

alt.Chart(df).transform_aggregate(
    mean_hp='mean(Horsepower)', groupby=['Origin']
).mark_bar().encode(
    x='Origin',
    y='mean_hp'
)
```

## 🧩 Interactividad — selecciones y filtros dinámicos


Altair soporta **selecciones interactivas** mediante `selection_point()` o `selection_interval()`.

📘 Ejemplo: selección por país con clic o leyenda.


```python

selector = alt.selection_point(fields=["Origin"], bind="legend")

alt.Chart(df).mark_circle(size=70).encode(
    x="Horsepower",
    y="Miles_per_Gallon",
    color=alt.condition(selector, "Origin:N", alt.value("lightgray"))
).add_params(selector).properties(title="Selección interactiva por origen")
```

## 🧱 facet() — Subgráficos por categoría


`facet()` permite dividir el gráfico en paneles según una variable categórica.

📘 Ejemplo: un gráfico por país de origen.


```python

alt.Chart(df).mark_point().encode(
    x="Horsepower",
    y="Miles_per_Gallon"
).facet(
    column="Origin"
).properties(title="Facetas por origen")
```

## 💾 Guardar gráfico


Puedes exportar gráficos a **HTML**, **PNG** o **SVG**.

📘 Ejemplo:


```python

chart = alt.Chart(df).mark_bar().encode(
    x="Origin",
    y="count()"
)
chart.save("grafico_origen.html")
print("Gráfico guardado como grafico_origen.html")
```
