# Introducción a spaCy


En este notebook aprenderemos a instalar spaCy, cargar un modelo en español y realizar tokenización básica.

```
import spacy
nlp = spacy.load('es_core_news_sm')
text = "Hola, me llamo Ana y vivo en Madrid."
doc = nlp(text)
for token in doc:
    print(token.text, token.pos_, token.dep_)
```
