Incluye creación de DataFrames, selección/filtrado y comparación de sintaxis con pandas.

```python

import numpy as np, pandas as pd
try:
    import polars as pl
    HAS_POLARS = True
    print("✅ Polars importado:", pl.__version__)
except Exception as e:
    HAS_POLARS = False
    print("⚠️ Polars no disponible en este entorno:", e)

rng = np.random.default_rng(0)
n = 1000
pdf = pd.DataFrame({
    "id": np.arange(n),
    "edad": rng.integers(18, 70, size=n),
    "ingresos": rng.normal(26000, 9000, size=n).round(2),
    "ciudad": rng.choice(["Cádiz","Sevilla","Málaga","Córdoba"], size=n),
    "compra": rng.choice([0,1], size=n, p=[0.45, 0.55])
})
pdf.head()
```
